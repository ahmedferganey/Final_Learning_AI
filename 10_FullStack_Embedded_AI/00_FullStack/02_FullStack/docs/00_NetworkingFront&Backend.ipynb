### 🚀 Python 3: Async/Await, Typing & Context Managers

Modern Python (3.7+) includes powerful features that make it ideal for building scalable, asynchronous web backends. FastAPI builds on top of these capabilities to offer high-performance request handling. Below is a breakdown of the essential Python features that power asynchronous behavior in FastAPI applications.


#### ✨ `async def`: Declaring Coroutines

Functions declared using `async def` define *coroutines*, which are special functions that support asynchronous operations. Unlike regular functions, coroutines allow pausing execution to perform non-blocking I/O, such as reading from a database or making a network call.

These functions do not return values directly; instead, they return a coroutine object, which needs to be awaited to retrieve its result.

```python
async def greet_user(name: str) -> str:
    return f"Hello, {name}!"

# Usage
import asyncio

result = asyncio.run(greet_user("Ahmed"))
print(result)  # Output: Hello, Ahmed!
```

Coroutines enable your application to handle many tasks concurrently with minimal thread usage.



#### ⏳ `await`: Pausing Execution

The `await` keyword is used *within an `async` function* to call another coroutine and wait for its result. While waiting, the event loop can continue running other tasks, making efficient use of resources.

You can only use `await` inside an `async def` function.

```python
async def get_user_data():
    return {"name": "Ahmed"}

async def display_user():
    user = await get_user_data()
    print(user)

# Usage
import asyncio
asyncio.run(display_user())
```

In this example, `display_user` pauses at `await get_user_data()` until the data is available.



#### ✍️ Typing: Type Hints for Better Development

Python typing adds static type annotations to improve development experience. These annotations:

* Help IDEs provide better autocompletion and documentation.
* Enable FastAPI to validate request and response models.
* Reduce runtime errors by catching mistakes during development.

```python
from typing import List, Dict

def get_names() -> List[str]:
    return ["Alice", "Bob"]

# Annotating function parameters and return values

def multiply(x: int, y: int) -> int:
    return x * y

# Annotating variables
person: Dict[str, str] = {"name": "Ahmed"}
```

FastAPI automatically converts these hints into OpenAPI docs and request validators.



#### 🔑 `async with`: Asynchronous Context Managers

`async with` is the asynchronous version of `with`, used for managing resources that support asynchronous entry and exit methods. Common use cases include:

* Database sessions
* Network connections (HTTP, sockets)
* File operations (in some async file libraries)

```python
import aiohttp

async def download_html(url: str) -> str:
    async with aiohttp.ClientSession() as session:
        async with session.get(url) as response:
            return await response.text()

# Usage
import asyncio
html = asyncio.run(download_html("https://example.com"))
print(html[:100])
```

Using `async with` ensures that resources are properly opened and closed, even if exceptions occur.



### 📝 Practical Example

Here’s a full example demonstrating `async def`, `await`, typing, and `async with`:

```python
from typing import Dict
import httpx

async def fetch_data(url: str) -> Dict:
    async with httpx.AsyncClient() as client:
        response = await client.get(url)
        return response.json()

# Usage
import asyncio
url = "https://jsonplaceholder.typicode.com/todos/1"
data = asyncio.run(fetch_data(url))
print(data)
```

In this example:

* `async def` defines the coroutine `fetch_data`.
* `await` waits for the HTTP GET request to complete.
* `async with` ensures the `httpx.AsyncClient` is properly opened and closed.
* The function return type is annotated as a `Dict` for clarity and correctness.
* `asyncio.run` is used to execute the coroutine in a synchronous environment.

FastAPI uses Python's async model to:

* Integrate easily with event loops, background tasks, and async libraries (e.g., `httpx`, `asyncpg`, `aiofiles`)

Understanding these core language features is essential for writing robust, clean, and performant FastAPI applications.

### ⚡ FastAPI Core Concepts

FastAPI is designed with developer experience and performance in mind. It offers a simple, powerful way to define API behavior using modern Python features. Let's break down the three foundational concepts: **Routing**, **Middleware**, and **Dependency Injection**, with clear explanations and practical usage examples.


#### ▶ Routing

**Routing** in FastAPI is handled using decorators like `@app.get`, `@app.post`, etc. These decorators define which function should handle a given HTTP method and path.

Each route is an asynchronous or synchronous function that receives request data and returns a response.

```python
from fastapi import FastAPI

app = FastAPI()

@app.get("/products")
async def get_products():
    return [{"id": 1, "name": "Widget"}]
```

* `@app.get("/products")`: This registers a GET route at `/products`.
* `get_products()`: This function handles the request.
* `async def`: Indicates it's an asynchronous function, enabling non-blocking behavior.
* Returns a list of product dictionaries as JSON.

You can also include parameters in routes:

```python
@app.get("/products/{product_id}")
async def get_product(product_id: int):
    return {"id": product_id, "name": "Sample Product"}
```


#### ▶ Middleware

**Middleware** in FastAPI lets you intercept requests and responses to add custom behavior. It can be used for logging, authentication, error handling, or modifying request/response data.

One common middleware is `CORSMiddleware`, which manages cross-origin resource sharing (CORS).

```python
from fastapi import FastAPI
from fastapi.middleware.cors import CORSMiddleware

app = FastAPI()

app.add_middleware(
    CORSMiddleware,
    allow_origins=["*"],        # Allow requests from any origin
    allow_credentials=True,     # Include credentials (cookies, headers)
    allow_methods=["*"],        # Allow all HTTP methods
    allow_headers=["*"],        # Allow all headers
)
```

You can also create **custom middleware** using the `@app.middleware("http")` decorator:

```python
@app.middleware("http")
async def log_requests(request, call_next):
    print(f"Incoming request: {request.method} {request.url}")
    response = await call_next(request)
    print(f"Response status: {response.status_code}")
    return response
```

This middleware logs the request method and URL before passing the request along, then logs the response status after it's returned.


#### ▶ Dependency Injection

**Dependency Injection** in FastAPI lets you define reusable components or logic (e.g., database sessions, authentication) and "inject" them into your route functions using the `Depends()` function.

This improves code modularity, testability, and reusability.

```python
from fastapi import Depends, FastAPI

app = FastAPI()

# Dependency function
def get_token():
    return "secure-token"

@app.get("/info")
async def read_info(token: str = Depends(get_token)):
    return {"token": token}
```

You can also use `yield` for dependencies that require setup/teardown, such as database sessions:

```python
from fastapi import Depends
from sqlalchemy.orm import Session

# Assuming SessionLocal is your database session factory
async def get_db():
    db = SessionLocal()
    try:
        yield db
    finally:
        db.close()

@app.get("/secure-data")
async def secure_data(db: Session = Depends(get_db)):
    return db.query(Models.Data).all()
```

* `get_db()` creates and yields a database session.
* `Depends(get_db)` injects the session into the route.
* FastAPI handles the lifecycle, calling `db.close()` when the request is done.


These core features of FastAPI provide a solid, scalable foundation for building modern APIs. They support clean, maintainable code that scales well in production environments.

### 🔐 Firebase JWT Validation in FastAPI

Firebase Authentication uses JSON Web Tokens (JWTs) to identify and authorize users securely. FastAPI can integrate Firebase JWT validation to secure endpoints using the `firebase_admin` SDK and FastAPI’s built-in security mechanisms.


#### 🔧 Step-by-Step Integration

##### ✅ 1. Initialize Firebase Admin SDK

You need a Firebase service account key in JSON format to authenticate with Firebase.

```python
import firebase_admin
from firebase_admin import auth, credentials

cred = credentials.Certificate("firebase-adminsdk.json")
firebase_admin.initialize_app(cred)
```

##### ✅ 2. Use FastAPI's HTTPBearer Security Scheme

FastAPI’s `HTTPBearer` extracts the bearer token from the request's `Authorization` header.

```python
from fastapi.security import HTTPBearer, HTTPAuthorizationCredentials
from fastapi import Security, HTTPException

security = HTTPBearer()
```

##### ✅ 3. Validate Firebase JWT Token

The token received is verified using Firebase Admin SDK's `verify_id_token()` method. If valid, the token is decoded and user info is returned. If invalid, a 401 Unauthorized error is raised.

```python
async def get_current_user(credentials: HTTPAuthorizationCredentials = Security(security)):
    try:
        token = credentials.credentials
        decoded_token = auth.verify_id_token(token)
        return decoded_token
    except Exception:
        raise HTTPException(status_code=401, detail="Invalid token")
```


#### 🔒 Example: Secure Route with Firebase Authentication

```python
from fastapi import FastAPI, Depends

app = FastAPI()

@app.get("/profile")
async def read_user_profile(user=Depends(get_current_user)):
    return {"user_id": user["uid"], "email": user.get("email")}
```

This ensures that only authenticated users with a valid Firebase token can access `/profile`.


#### 🧪 Testing the Endpoint

Use an authenticated Firebase client (e.g., Firebase JS SDK) to sign in and obtain the token, then include it in the `Authorization` header as:

```
Authorization: Bearer <your_id_token>
```


#### 📌 Notes

* Tokens are time-limited; handle token expiration and re-authentication on the client side.
* You can extend `get_current_user` to verify roles or claims.
* Avoid re-initializing Firebase multiple times; do it once at startup.


Firebase JWT validation adds a robust security layer to your FastAPI backend and enables seamless integration with mobile/web apps using Firebase Authentication.

### 🔧 RESTful API Design: Products, Stores, Orders

When building backend systems that manage products, stores, and customer orders, it's important to design clean and intuitive RESTful APIs. This ensures your service is scalable, easy to integrate with, and aligns with standard HTTP methods.


#### 🧱 Resource-Oriented Endpoints

* `/products`: Manage the product catalog

  * `GET /products`: List all products
  * `POST /products`: Add a new product
  * `GET /products/{product_id}`: Get a specific product
  * `PUT /products/{product_id}`: Update a product
  * `DELETE /products/{product_id}`: Delete a product

* `/stores`: Manage store locations and their inventories

  * `GET /stores`: List all stores
  * `POST /stores`: Add a new store
  * `GET /stores/{store_id}`: Get store details
  * `PUT /stores/{store_id}`: Update store information
  * `DELETE /stores/{store_id}`: Remove a store

* `/orders`: Handle customer orders

  * `GET /orders`: List all orders
  * `POST /orders`: Create a new order
  * `GET /orders/{order_id}`: Get order details
  * `PUT /orders/{order_id}`: Update an order
  * `DELETE /orders/{order_id}`: Cancel/delete an order


#### ⚙️ HTTP Semantics in Action

* `GET`: Retrieve resources (e.g., list of products, single order)
* `POST`: Create new resources (e.g., new product or order)
* `PUT`: Full update of a resource
* `PATCH`: Partial update (e.g., change order status)
* `DELETE`: Remove a resource permanently


#### 📌 Best Practices

* Use plural nouns (`/products`, `/stores`, `/orders`)
* Use nested routes for related resources (e.g., `/stores/{store_id}/products`)
* Return appropriate HTTP status codes (e.g., 200 OK, 201 Created, 404 Not Found, 400 Bad Request)
* Keep URLs resource-focused and avoid action words (e.g., avoid `/getProducts`, prefer `/products`)


This approach to RESTful API design ensures your system is predictable, easy to consume, and aligned with industry standards.

### ✅ Pydantic: Validation & Response Models

Pydantic is a data validation and parsing library used extensively with FastAPI to ensure that incoming and outgoing data conforms to the expected structure. It uses Python type annotations to automatically enforce data types, validate input, and generate example schemas for documentation.


#### 📥 Defining Input and Output Models

You define models by subclassing `BaseModel`. Each field is declared with a type, and optional fields can use `Optional[]` or provide default values.

```python
from pydantic import BaseModel, Field
from typing import Optional

class ProductIn(BaseModel):
    name: str = Field(..., min_length=1, example="Wireless Mouse")
    price: float = Field(..., gt=0, example=29.99)
    description: Optional[str] = Field(None, example="2.4GHz wireless optical mouse")

class ProductOut(ProductIn):
    id: int = Field(..., example=101)
```

The `Field(...)` function allows for additional validation rules (e.g., `min_length`, `gt` for greater than) and metadata like `example` used in API documentation.


#### 🔁 Usage in FastAPI Endpoints

Use the models as input (`request body`) and output (`response_model`) in your endpoints:

```python
from fastapi import FastAPI

app = FastAPI()

@app.post("/products", response_model=ProductOut)
async def create_product(product: ProductIn):
    # Simulate product creation
    return {"id": 1, **product.dict()}
```

* `product: ProductIn` automatically parses and validates the request body.
* `response_model=ProductOut` ensures only defined fields are returned and validated.


#### ✅ Benefits of Using Pydantic in FastAPI

* **Automatic validation**: Ensures all data is correct before processing.
* **Cleaner code**: Eliminates boilerplate validation logic.
* **Interactive docs**: Models enhance FastAPI's autogenerated Swagger and ReDoc UI.
* **Serialization & deserialization**: Transforms data between JSON and Python types seamlessly.


#### 🔍 Advanced Features

* **Validators**: Add custom validation logic using `@validator` decorators.
* **Nested models**: Structure complex data with nested `BaseModel` classes.
* **Aliases and renaming**: Handle snake\_case to camelCase transformations easily.

Pydantic plays a crucial role in building reliable, well-documented, and maintainable FastAPI applications by making data handling explicit and safe.

### 📃 SQLAlchemy ORM: Models, Queries, Migrations

SQLAlchemy is a powerful SQL toolkit and Object Relational Mapper (ORM) for Python. It enables developers to map Python classes to database tables and interact with databases using high-level Python code instead of raw SQL.


#### ▶ Models

Define models by creating classes that inherit from a declarative base. Each attribute corresponds to a column in the database:

```python
from sqlalchemy import Column, Integer, String
from sqlalchemy.orm import declarative_base

Base = declarative_base()

class Product(Base):
    __tablename__ = "products"
    id = Column(Integer, primary_key=True)
    name = Column(String)
    price = Column(Integer)
```

* `__tablename__` defines the table name in the database.
* Columns are defined using SQLAlchemy types (`Integer`, `String`, etc.).


#### ▶ Async Queries

With SQLAlchemy 2.0's native async support (or using the `databases` library), you can interact with the database asynchronously:

```python
from sqlalchemy.ext.asyncio import AsyncSession
from sqlalchemy.future import select

async with AsyncSession(engine) as session:
    result = await session.execute(select(Product))
    products = result.scalars().all()
```

* `AsyncSession` is used for asynchronous interactions.
* `select(Product)` builds a query to fetch all rows from the `products` table.
* `scalars().all()` extracts and returns the list of `Product` objects.

#### ▶ Migrations with Alembic

Alembic is a lightweight database migration tool for SQLAlchemy. It tracks schema changes and applies them to the database.

**Steps to use Alembic:**

```bash
alembic init alembic                 # Initializes Alembic in your project
alembic revision --autogenerate -m "init"  # Creates a new migration script
alembic upgrade head                 # Applies the migration to the database
```

* Alembic scans your models and generates migration scripts.
* `alembic upgrade head` applies the latest changes.

---

SQLAlchemy ORM simplifies database access and management, making it easier to build scalable, maintainable backend applications with Python.

### 🐇 Emitting Events to RabbitMQ

#### ▶ Using aio-pika

```python
import aio_pika

async def send_event(message: dict):
    connection = await aio_pika.connect_robust("amqp://guest:guest@localhost/")
    async with connection:
        channel = await connection.channel()
        await channel.default_exchange.publish(
            aio_pika.Message(body=json.dumps(message).encode()),
            routing_key="events"
        )
```

#### ▶ Using kombu

```python
from kombu import Connection, Exchange, Producer

conn = Connection("amqp://guest:guest@localhost//")
exchange = Exchange("events", type="direct")

with conn:
    producer = Producer(conn)
    producer.publish(
        {"event": "order_created"},
        exchange=exchange,
        routing_key="events"
    )
```
