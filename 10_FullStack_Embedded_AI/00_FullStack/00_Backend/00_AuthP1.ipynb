## 🔐 Authentication Foundations

---

## 1. Foundations

This section establishes the **core mental models** required to understand authentication systems in real-world architectures.

---

## 1.1 Authentication vs Authorization

### 🔹 Authentication (AuthN)

**Definition:**
Verifying *who you are*.

### Examples:

* Logging in with email/password
* Providing a JWT token
* Using OAuth login (Google, GitHub)

### Result:

➡️ System identifies the user

---

### 🔹 Authorization (AuthZ)

**Definition:**
Determining *what you are allowed to do*.

### Examples:

* Admin can delete users
* User can only view their own data
* Service can access specific API

### Result:

➡️ System grants permissions

---

### 🔥 Key Difference

| Aspect        | Authentication | Authorization          |
| ------------- | -------------- | ---------------------- |
| Question      | Who are you?   | What can you do?       |
| Happens First | ✅ Yes          | ❌ After authentication |
| Output        | Identity       | Permissions            |

👉 Flow:

Client → Authentication → Authorization → Resource Access

---

## 1.2 Stateful vs Stateless Authentication

---

### 🔹 Stateful Authentication

**Definition:**
Server stores user session state.

### How it works:

1. User logs in
2. Server creates session (stored in memory/DB)
3. Client receives session ID (cookie)
4. Server validates session on each request

### Characteristics:

* Server holds state
* Requires session storage
* Easier to revoke sessions

### Example:

* Traditional web apps (PHP, Django sessions)

---

### 🔹 Stateless Authentication

**Definition:**
Server does NOT store session; client sends all required data each request.

### How it works:

1. User logs in
2. Server returns token (JWT)
3. Client sends token with each request
4. Server validates token (no DB lookup needed)

### Characteristics:

* No server-side session storage
* Scales easily (distributed systems)
* Harder to revoke tokens

### Example:

* Modern APIs (JWT-based auth)

---

### 🔥 Comparison

| Feature       | Stateful         | Stateless          |
| ------------- | ---------------- | ------------------ |
| Server Memory | Required         | Not required       |
| Scalability   | Limited          | High               |
| Revocation    | Easy             | Hard               |
| Performance   | DB lookup needed | Faster (no lookup) |

---

## 1.3 Sessions vs Tokens

---

### 🔹 Sessions

**Definition:**
Server-side stored authentication state.

### Flow:

1. User logs in
2. Server creates session
3. Client stores session ID (cookie)
4. Server validates session ID

### Characteristics:

* Stored on server
* Requires session store (Redis/DB)
* Tight coupling to server

---

### 🔹 Tokens

**Definition:**
Self-contained authentication data sent by client.

### Flow:

1. User logs in
2. Server issues token (JWT)
3. Client stores token
4. Token sent with each request

### Characteristics:

* Stored on client
* Contains encoded user data
* No server storage required

---

### 🔥 Comparison

| Feature     | Sessions                | Tokens (JWT)         |
| ----------- | ----------------------- | -------------------- |
| Storage     | Server-side             | Client-side          |
| Scalability | Harder                  | Easier               |
| Security    | Strong (server control) | Depends on handling  |
| Use Case    | Web apps                | APIs / microservices |

---

## 🚀 Key Takeaways

* Authentication = identity, Authorization = permissions
* Stateful = server remembers you
* Stateless = client proves identity every request
* Sessions = server-side
* Tokens = client-side

👉 Modern systems (APIs, microservices, AI platforms) rely heavily on:

* Stateless authentication
* Token-based systems (JWT)



# 2. HTTP & Security Basics


## 2. 🌐 HTTP & Security Basics (Authentication Context)


This section explains the **network and protocol-level foundations** that make authentication possible.

---

## 2.1 HTTP Request Lifecycle

### 🔹 Definition

The sequence of steps that occur when a client (browser/app) sends a request to a server and receives a response.

---

### 🔹 Full Lifecycle

1. **Client Request Initiation**

   * User triggers request (click, API call)

2. **DNS Resolution**

   * Domain → IP address

3. **TCP Connection**

   * Establish connection with server

4. **TLS Handshake (if HTTPS)**

   * Secure encrypted channel setup

5. **HTTP Request Sent**

   * Includes method, headers, body

6. **Server Processing**

   * Authentication check
   * Authorization check
   * Business logic

7. **HTTP Response Returned**

   * Status code + data

---

### 🔥 Where Authentication Happens

➡️ At the **server processing stage**

* Server reads headers (Authorization)
* Validates credentials/token
* Decides access

---

### 🔹 Example Request

```http
GET /api/data HTTP/1.1
Host: example.com
Authorization: Bearer <token>
```

---

## 2.2 HTTP Headers & Authorization

---

### 🔹 What are HTTP Headers?

Metadata sent with requests and responses.

---

### 🔹 Important Headers for Authentication

#### 1. Authorization Header

```http
Authorization: Bearer <token>
```

Used for:

* JWT tokens
* OAuth tokens

---

#### 2. API Key Header

```http
x-api-key: YOUR_KEY
```

---

#### 3. Cookie Header (Sessions)

```http
Cookie: session_id=abc123
```

---

### 🔥 Common Auth Header Formats

| Method       | Header Example                         |
| ------------ | -------------------------------------- |
| Bearer Token | Authorization: Bearer token            |
| Basic Auth   | Authorization: Basic base64(user:pass) |
| API Key      | x-api-key: key                         |

---

### 🔹 Why Headers?

* Standardized
* Secure over HTTPS
* Decoupled from request body

---

## 2.3 HTTPS & TLS

---

### 🔹 Problem Without HTTPS

HTTP is **plain text**:

➡️ Anyone can intercept:

* Passwords
* Tokens
* API keys

---

### 🔹 Solution: HTTPS (HTTP + TLS)

Encrypts communication between client and server.

---

### 🔹 TLS Handshake (Simplified)

1. Client sends request to server
2. Server sends certificate (public key)
3. Client verifies certificate
4. Session key is generated
5. Encrypted communication begins

---

### 🔥 What TLS Protects

* Authorization headers
* Cookies
* Request/response data

---

### 🔹 Important Concepts

#### Encryption

* Data is unreadable without key

#### Certificates

* Verify server identity

#### Public / Private Key

* Used to establish secure channel

---

### 🔹 Why HTTPS is Mandatory

| Without HTTPS        | With HTTPS           |
| -------------------- | -------------------- |
| Data visible         | Data encrypted       |
| Easy to attack       | Secure communication |
| Tokens can be stolen | Tokens protected     |

---

## 🚀 Key Takeaways

* HTTP lifecycle defines **where auth happens**
* Headers carry **authentication data**
* HTTPS ensures **security of credentials**

👉 Without understanding this layer, authentication systems cannot be designed securely.



# 3. Authentication Methods




This section covers the **core authentication mechanisms used in modern systems**, from simple to production-grade.

---

## 3.1 Bearer Token Authentication

### 🔹 Definition

A token-based authentication method where the client sends a token in the HTTP header.

```http
Authorization: Bearer <token>
```

---

### 3.1.1 JWT Structure (Header, Payload, Signature)

A JWT (JSON Web Token) consists of 3 parts:

```
HEADER.PAYLOAD.SIGNATURE
```

#### Header

* Algorithm (e.g., HS256)

#### Payload

* User data (claims)
* Example:

```json
{
  "user_id": 123,
  "role": "admin",
  "exp": 1712345678
}
```

#### Signature

* Ensures token integrity
* Generated using secret/private key

---

### 3.1.2 Token Expiration & Validation

#### Expiration

* Tokens include `exp` field
* Prevents long-term misuse

#### Validation Steps

1. Check signature
2. Check expiration
3. Validate claims (role, issuer)

---

### 3.1.3 JWT vs Sessions

| Feature     | JWT         | Sessions    |
| ----------- | ----------- | ----------- |
| Storage     | Client-side | Server-side |
| Scalability | High        | Limited     |
| Revocation  | Hard        | Easy        |
| Use Case    | APIs        | Web apps    |

---

## 3.2 API Key Authentication

---

### 🔹 Definition

A simple authentication mechanism using a static key.

---

### 3.2.1 API Key in Headers vs Query Params

#### Header (Recommended)

```http
x-api-key: YOUR_KEY
```

#### Query Param (Not Recommended)

```http
GET /data?api_key=123
```

👉 Headers are safer (not logged in URLs)

---

### 3.2.2 Security Tradeoffs

| Aspect     | API Key                |
| ---------- | ---------------------- |
| Security   | Medium                 |
| Complexity | Low                    |
| Identity   | Weak (no user context) |

👉 Best for simple or internal APIs

---

## 3.3 HTTP Basic Authentication

---

### 🔹 Definition

Uses username and password encoded in Base64.

```http
Authorization: Basic base64(username:password)
```

---

### 3.3.1 Base64 Encoding Explained

* Encoding, NOT encryption
* Easily decoded

Example:

```
user:pass → dXNlcjpwYXNz
```

---

### 3.3.2 Security Limitations

* Credentials sent every request
* No encryption without HTTPS
* Vulnerable to interception

👉 Only safe with HTTPS

---

## 3.4 OAuth2

---

### 🔹 Definition

A secure authorization framework for delegated access.

---

### 3.4.1 OAuth2 Concepts

* Resource Owner (User)
* Client (App)
* Authorization Server
* Resource Server (API)

---

### 3.4.2 Client Credentials Flow

Used for **service-to-service authentication**

#### Flow:

1. Client sends client_id + client_secret
2. Auth server returns access token
3. Client uses token

---

### 3.4.3 Authorization Code Flow

Used for **user login (e.g., Google login)**

#### Flow:

1. User redirected to auth server
2. User logs in
3. Server returns authorization code
4. Client exchanges code for token

---

### 3.4.4 Access Tokens vs Refresh Tokens

| Token Type    | Purpose                       |
| ------------- | ----------------------------- |
| Access Token  | Used for API requests         |
| Refresh Token | Used to get new access tokens |

---

## 🚀 Key Takeaways

* JWT (Bearer) = modern API standard
* API Keys = simple but limited
* Basic Auth = legacy, avoid in production
* OAuth2 = enterprise-grade authentication

👉 Production systems typically use:

* OAuth2 + JWT
* API Gateway enforcing authentication
