### **NLP | Text Preprocessing | Word Embedding**



**Word Embedding** is a technique used in Natural Language Processing (NLP) to represent words as dense vectors in a continuous vector space. Unlike traditional methods like Bag of Words (BoW) or TF-IDF, word embeddings capture semantic meanings and relationships between words, enabling models to understand context, synonyms, and analogies.


#### 5 Types of Word Embedding Techniques

There are two main categories of word embedding methods:

1. **Frequency-based Embedding**
2. **Prediction-based Embeddings**


##### Frequency-based Embeddings

1. **Term Frequency-Inverse Document Frequency (TF-IDF)**
2. **Co-occurrence Matrix**


##### Prediction-Based Embeddings

3. **Word2Vec (Skip-gram and Continuous Bag of Words)**
4. **FastText**
5. **GloVe (Global Vectors for Word Representation)**

![](https://miro.medium.com/v2/resize:fit:2000/format:webp/1*DpgA3bbFBmQ9I_amVDFvCQ.png)

### 1. **How Word Embeddings Work**

Word embeddings map words to vectors of fixed dimensions (usually between 50 and 300), where each word is represented by a point in a multi-dimensional space. Similar words (in meaning or usage) are mapped to nearby points in this space.

- **Training Word Embeddings**: Word embeddings are typically trained using large corpora of text through algorithms such as Word2Vec, GloVe, or FastText.
  
- **Word2Vec**: Trains embeddings using either the Skip-Gram model (predicts surrounding words from the target word) or the Continuous Bag of Words (CBOW) model (predicts the target word from the surrounding words).
  
- **GloVe**: Factorizes the word co-occurrence matrix of a corpus to find the most informative word vectors.


### 2. **Advantages of Word Embeddings**

- **Semantic Representation**: Word embeddings capture the semantic meaning of words, allowing words with similar meanings to have similar vector representations (e.g., "king" and "queen" might be close in the vector space).
  
- **Context Sensitivity**: Embeddings can account for context in which a word appears, unlike BoW or TF-IDF, which treat words independently of context.

- **Dimensionality Reduction**: Embeddings represent words in dense vectors (as opposed to sparse representations like BoW or One-Hot Encoding), leading to a more compact and efficient representation of words.


### 3. **Popular Word Embedding Models**

- **Word2Vec**: A shallow neural network model that learns word representations by predicting neighboring words in a text corpus. Commonly used for word analogy tasks, such as "king - man + woman = queen."
  
- **GloVe (Global Vectors for Word Representation)**: A matrix factorization-based technique that captures the global statistical information of a corpus. It performs well for tasks like document similarity.
  
- **FastText**: Extends Word2Vec by representing words as bags of character n-grams. This helps in capturing morphological information and is useful for languages with rich morphology or for handling out-of-vocabulary words.

- **ELMo (Embeddings from Language Models)**: Contextual embeddings generated by training a bidirectional LSTM on a large corpus. ELMo embeddings change depending on the context of the word in a sentence.

- **BERT (Bidirectional Encoder Representations from Transformers)**: A transformer-based model that produces contextual word embeddings, meaning the same word can have different representations depending on its context in the sentence.


### 4. **Applications of Word Embeddings**

- **Text Classification**: Word embeddings improve the performance of models by providing rich semantic features that can be used in classifiers.
  
- **Named Entity Recognition (NER)**: By capturing contextual relationships, embeddings help models identify and classify entities like names, dates, and locations in text.

- **Sentiment Analysis**: Word embeddings provide a better understanding of sentiment by capturing nuances in word meanings and context.

- **Word Similarity and Analogies**: Embeddings are widely used to compute word similarity, find analogies, and measure the relationship between words (e.g., "man" is to "woman" as "king" is to "queen").


### 5. **Challenges in Word Embeddings**

- **Out-of-Vocabulary (OOV) Words**: Word embeddings may not handle words that were not seen during training, which can be problematic for rare or new words.

- **Bias**: Word embeddings can inherit biases present in the data they are trained on. For example, embeddings may encode gender or racial biases, which can affect downstream tasks.

- **Contextual Representations**: While word embeddings capture semantic relationships, they still struggle with capturing fine-grained context, especially when words have multiple meanings (polysemy).


### 6. **Modern Alternatives and Advances**

- **Contextual Embeddings**: Models like BERT and GPT provide dynamic, context-aware embeddings that adjust based on the surrounding words, offering more accurate representations for words with multiple meanings.
  
- **Transformer Models**: Transformers like BERT, GPT, and T5 generate embeddings based on the entire sentence or document, capturing deeper linguistic and contextual understanding.

- **Pretrained Embeddings**: Many NLP tasks now use pretrained embeddings like BERT, RoBERTa, and GPT models, which have been trained on vast corpora and can be fine-tuned for specific tasks.


### 7. **Conclusion**

Word embeddings are a critical component in modern NLP pipelines, offering a more efficient and semantically rich way to represent words compared to traditional methods. While models like Word2Vec, GloVe, and FastText provide strong embeddings for tasks like classification and similarity, newer techniques like BERT and GPT are pushing the boundaries by providing dynamic, context-sensitive representations.


### **1. NLP | Text Preprocessing**

Text preprocessing is a critical step in Natural Language Processing (NLP) that involves preparing raw text data for analysis or machine learning models. Since real-world text is noisy and inconsistent, preprocessing helps to clean and standardize the data, ensuring that it is suitable for further analysis.

#### Key Steps in Text Preprocessing:

1. **Lowercasing**:
   - Convert all the text to lowercase to ensure uniformity, as "Text" and "text" are treated as different words if case is not normalized.
   - Example: `"Hello World!"` becomes `"hello world!"`

2. **Tokenization**:
   - The process of splitting text into smaller units, such as words, phrases, or sentences.
   - Example: `"I love NLP!"` becomes `["I", "love", "NLP"]`

3. **Removing Punctuation**:
   - Removing punctuation marks such as commas, periods, and exclamation points, as they often do not contribute to the meaning in text analysis.
   - Example: `"Hello, world!"` becomes `"Hello world"`

4. **Removing Stopwords**:
   - Stopwords are common words like "and", "the", "is", etc., that do not carry significant meaning and are removed to reduce noise.
   - Example: `"The quick brown fox"` becomes `"quick brown fox"`

5. **Stemming**:
   - Reducing words to their root form by removing suffixes. This helps to normalize the words so that they can be treated as the same word.
   - Example: `"running"`, `"ran"`, and `"runner"` all become `"run"`

6. **Lemmatization**:
   - Similar to stemming, but instead of chopping off suffixes, lemmatization reduces words to their base or dictionary form, using vocabulary and morphological analysis.
   - Example: `"better"` becomes `"good"`, `"running"` becomes `"run"`

7. **Removing Numbers**:
   - Often, numbers do not provide meaningful information in many NLP tasks, and thus they may be removed.
   - Example: `"There are 3 apples"` becomes `"There are apples"`

8. **Removing Special Characters**:
   - Stripping out non-alphabetic characters like `@`, `#`, and `$` to focus on meaningful words.
   - Example: `"I love @NLP!"` becomes `"I love NLP"`

9. **Handling Negations**:
   - Special attention is often given to negations (e.g., "not", "never"), as they can affect the meaning of a sentence.
   - Example: `"I don't like apples"` might be transformed into `"I like not apples"`

10. **Handling Unicode and Encoding Issues**:
    - Normalizing text to handle non-ASCII characters and ensuring text is properly encoded.
    - Example: `"café"` could be encoded as `"cafe"` to avoid issues with special characters.

#### Why Text Preprocessing is Important:
- **Data Quality**: Raw text data can be noisy, inconsistent, and full of irrelevant information, making it difficult to extract meaningful patterns.
- **Improved Model Performance**: Cleaned and preprocessed text improves the quality and accuracy of machine learning models.
- **Feature Engineering**: Text preprocessing can help extract valuable features like word frequency, n-grams, or named entities, which are useful for training models.

#### Tools for Text Preprocessing:
- **NLTK (Natural Language Toolkit)**: Provides functions for tokenization, removing stopwords, stemming, and more.
- **spaCy**: A library that offers fast, efficient preprocessing methods like tokenization, lemmatization, and named entity recognition.
- **Regex (Regular Expressions)**: A powerful tool for pattern matching and text cleaning tasks such as removing unwanted characters.
- **Gensim**: Useful for preprocessing text data when working with large corpora and topic modeling.

Text preprocessing is essential for ensuring that the input data is clean, structured, and ready for downstream tasks such as text classification, sentiment analysis, and machine translation.


#### **1.1 NLP | Text Preprocessing | Tokenization**

**Tokenization** is one of the fundamental steps in natural language processing (NLP). It involves splitting text into smaller units called **tokens**. These tokens can be words, subwords, or characters, depending on the approach used. Tokenization serves as the foundation for many NLP tasks such as text classification, machine translation, and named entity recognition (NER).

#### 1. **Types of Tokenization**

- **Word Tokenization**: This method splits a sentence into individual words. It's the most common form of tokenization.
  - Example:
    - Input: `"I love programming"`
    - Tokens: `["I", "love", "programming"]`

- **Subword Tokenization**: Subword tokenization breaks down words into smaller units, which is especially useful for handling rare or out-of-vocabulary words. This is a common approach used by models like **Byte Pair Encoding (BPE)** and **FastText**.
  - Example:
    - Input: `"unhappiness"`
    - Tokens: `["un", "happiness"]`

- **Character Tokenization**: In this method, text is split into individual characters. This approach is useful when dealing with languages with complex morphology or when modeling at a very fine-grained level.
  - Example:
    - Input: `"love"`
    - Tokens: `["l", "o", "v", "e"]`

![](https://i0.wp.com/eastgate-software.com/wp-content/uploads/2024/08/Blogs-photos-2.jpg?resize=1140%2C642&ssl=1)

#### 2. **Why Tokenization is Important**

Tokenization is crucial because:
- It allows the model to work with numerical representations of text.
- It breaks down text into manageable pieces for downstream tasks.
- Tokenization helps in handling variations in words and allows models to recognize patterns in text, such as **word stems** or **subword units**.

#### 3. **Tokenization in Python**

Python provides several libraries for tokenization:
- **nltk (Natural Language Toolkit)**:
  - The `nltk.word_tokenize()` function splits text into words using a pre-trained model.
  - Example:
    ```python
    import nltk
    nltk.download('punkt')
    text = "I love programming."
    tokens = nltk.word_tokenize(text)
    print(tokens)  # Output: ['I', 'love', 'programming', '.']
    ```
  
- **spaCy**:
  - spaCy is another popular NLP library that provides fast tokenization. It's more efficient than nltk and is often used in production systems.
  - Example:
    ```python
    import spacy
    nlp = spacy.load('en_core_web_sm')
    text = "I love programming."
    doc = nlp(text)
    tokens = [token.text for token in doc]
    print(tokens)  # Output: ['I', 'love', 'programming', '.']
    ```

- **Hugging Face Tokenizers**:
  - Hugging Face provides state-of-the-art tokenizers used with transformer models like BERT and GPT.
  - Example:
    ```python
    from transformers import AutoTokenizer
    tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')
    text = "I love programming."
    tokens = tokenizer.tokenize(text)
    print(tokens)  # Output: ['i', 'love', 'programming']
    ```

#### 4. **Challenges in Tokenization**

- **Handling Punctuation**: Deciding whether to treat punctuation as separate tokens or as part of surrounding words can be challenging.
  - Example: `"I love programming!"` could result in two tokens `["I", "love", "programming", "!"]` or four tokens `["I", "love", "programming", "!"]`.
  
- **Compound Words and Hyphenated Terms**: Some words might be hyphenated or combined, and deciding whether to split them or keep them intact depends on the task.
  - Example: `"high-tech"` could be treated as a single token or as two tokens `["high", "tech"]`.

- **Language-Specific Issues**: Some languages, like Chinese or Japanese, don't have clear word boundaries, making tokenization more complex.

#### 5. **Advanced Tokenization Techniques**

- **Byte Pair Encoding (BPE)**: BPE splits words into subwords based on the frequency of character pairs, helping handle rare words. This method is used in models like **GPT-2**.
  
- **WordPiece**: Similar to BPE, WordPiece splits words into subword units, but it focuses on maximizing the likelihood of the training data.

- **SentencePiece**: SentencePiece is an unsupervised text tokenizer and detokenizer that treats the text as a sequence of Unicode characters and uses subword units to build a vocabulary.

#### 6. **Conclusion**

Tokenization is a critical preprocessing step in NLP that helps convert raw text into meaningful units for further analysis. Whether it's word, subword, or character tokenization, the choice of tokenization technique will depend on the specific task and the characteristics of the language being processed. Understanding how to properly tokenize text will significantly improve the performance of NLP models and ensure that they can handle different kinds of text data.


````python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')
text = "I love programming."
tokens = tokenizer.tokenize(text)
print(tokens)  # Output: ['i', 'love', 'programming','.']
````

#### **1.3 NLP | Text Preprocessing | Stemming**

**Stemming** is a text normalization technique used in natural language processing (NLP) to reduce words to their root form. The process involves chopping off prefixes and suffixes from words to obtain their base form, known as a **stem**. Stemming is useful for tasks where it is important to group words with similar meanings but different forms (e.g., "running", "runner", "ran" → "run").

Unlike **lemmatization**, which considers the context of a word and converts it to its dictionary form (lemma), stemming typically produces non-dictionary words by removing affixes in a mechanical way.

![](https://wisdomml.in/wp-content/uploads/2022/08/stem_feats.jpg)

#### 1. **Why Stemming is Important**

- **Reduces Variations**: Stemming reduces words with the same root meaning into a single form, which helps improve the efficiency and accuracy of NLP models, especially in tasks like text classification and information retrieval.
- **Simplifies Text**: By converting words to their root form, stemming simplifies text, making it easier for algorithms to process large datasets.
- **Reduces Feature Space**: It minimizes the number of unique tokens (features) by mapping various word forms to a common root.

#### 2. **Popular Stemming Algorithms**

- **Porter Stemmer**: One of the most widely used stemming algorithms, which uses a series of rules to strip off common suffixes in English.
  - Example:
    - Input: `"running"`, `"runner"`, `"ran"`
    - Output: `"run"`, `"run"`, `"ran"`

- **Lancaster Stemmer**: A more aggressive stemming algorithm than the Porter Stemmer. It tends to cut off more characters and produces shorter stems.
  - Example:
    - Input: `"running"`, `"runner"`, `"ran"`
    - Output: `"run"`, `"run"`, `"ran"`

- **Snowball Stemmer**: Also known as the "English Snowball Stemmer", it is an improved version of the Porter Stemmer with better accuracy and more efficient performance.
  - Example:
    - Input: `"running"`, `"runner"`, `"ran"`
    - Output: `"run"`, `"run"`, `"ran"`

#### 3. **Stemming in Python**

Python libraries like **nltk** provide built-in implementations of various stemming algorithms.

- **Using NLTK’s Porter Stemmer**:

```python
  import nltk
  from nltk.stem import PorterStemmer

  nltk.download('punkt')
  ps = PorterStemmer()

  words = ["running", "runner", "ran"]
  stemmed_words = [ps.stem(word) for word in words]
  print(stemmed_words)  # Output: ['run', 'runner', 'ran']
```


- **Using NLTK’s Lancaster Stemmer**:

```python
from nltk.stem import LancasterStemmer
lancaster_stemmer = LancasterStemmer()
words = ["running", "runner", "ran"]
stemmed_words = [lancaster_stemmer.stem(word) for word in words]
print(stemmed_words)  # Output: ['run', 'run', 'ran']

```

#### **4. Challenges in Stemming**

- **Overstemming**: This occurs when the stemming algorithm removes too many characters, leading to stems that are not meaningful or accurate. For example, "unhappiness" may be reduced to "unhappi", which may not have a useful meaning.
- **Understemming**: This happens when the algorithm fails to reduce words to their root form. For example, the algorithm might fail to stem both "running" and "runner" to the same root word.
- **Irregular Words**: Stemming does not handle irregular words well, like "better" and "good" (which are unrelated but share a meaning), leading to a loss of important information.

#### **5. Stemming vs Lemmatization**

- **Stemming**: Reduces words to a root form by removing suffixes. It is faster but can result in stems that are not valid words.
- **Lemmatization**: Reduces words to their dictionary form (lemma) based on context and part of speech. It produces more meaningful results but is slower than stemming.

#### **6. Advanced Stemming Techniques**

- **Stemming with Context**: Some models use context-aware stemming, combining aspects of stemming with more sophisticated techniques like part-of-speech tagging to decide whether to stem a word.
- **Hybrid Methods**: Some modern systems use a combination of lemmatization and stemming to balance between speed and accuracy.

#### **7. Conclusion**

Stemming is a powerful technique to reduce words to their root forms and is commonly used in NLP to improve model performance, especially in tasks like information retrieval and text classification. While stemming may lose some subtle nuances of language, it helps simplify and standardize text data. However, one should be cautious of overstemming and understemming, which could affect the quality of the results in certain tasks.


````python
from nltk.stem import LancasterStemmer
lancaster_stemmer = LancasterStemmer()
words = ["running", "runner", "ran"]
stemmed_words = [lancaster_stemmer.stem(word) for word in words]
print(stemmed_words)  # Output: ['run', 'run', 'ran']
````

#### **1.4 NLP | Text Preprocessing | Lemmatization**

**Lemmatization** is a text normalization technique used in natural language processing (NLP) to reduce words to their dictionary form, known as a **lemma**. Unlike stemming, lemmatization considers the context and meaning of a word, producing more accurate results.

For example:
- Words like `"running"`, `"ran"`, and `"runner"` are reduced to the lemma `"run"`.
- Words like `"better"` and `"good"` are normalized to `"good"` based on their meaning.


### 1. **Why Lemmatization is Important**
- **Context Awareness**: Lemmatization uses part-of-speech tagging and linguistic rules, ensuring that the output retains its grammatical correctness.
- **Accuracy**: It provides meaningful base forms, unlike stemming, which may produce truncated or non-dictionary words.
- **Enhances NLP Models**: Lemmatization improves text quality for tasks like sentiment analysis, text classification, and search engines.


### 2. **How Lemmatization Works**

Lemmatization requires **linguistic resources**, such as a **dictionary** or **lexicon**, to map words to their base forms. The process involves:
- Identifying the **part of speech** (POS) of a word.
- Applying the rules to convert the word to its lemma.


### 3. **Lemmatization in Python**

Popular NLP libraries like **NLTK** and **spaCy** provide tools for lemmatization.

#### **Using NLTK**
```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
import nltk

nltk.download('wordnet')
nltk.download('omw-1.4')

lemmatizer = WordNetLemmatizer()

words = ["running", "ran", "runner", "better"]
lemmas = [lemmatizer.lemmatize(word, pos=wordnet.VERB) for word in words]
print(lemmas)  # Output: ['run', 'run', 'run', 'better']
```

#### **Using spaCy**
```python
import spacy

nlp = spacy.load("en_core_web_sm")
text = "running ran runner better"
doc = nlp(text)

lemmas = [token.lemma_ for token in doc]
```

- **Ambiguity**: Words with multiple meanings (e.g., "bark") may lead to incorrect lemmatization without sufficient context.


#### **6. Modern Alternatives**

- **Neural Lemmatizers**: Use machine learning models to predict lemmas based on context.
- **Hybrid Methods**: Combine lemmatization with stemming for optimized speed and accuracy.


#### **7. Conclusion**

Lemmatization is a crucial step in text preprocessing, providing cleaner and more meaningful text for NLP applications. Although it is slower than stemming, its accuracy and context-awareness make it a preferred choice for many text analysis tasks.

### **2. NLP | Text Preprocessing | BoW, TF-IDF, Unigrams & Bigrams**

Text preprocessing plays an essential role in transforming raw text into a structured format suitable for machine learning models. In this section, we will focus on the techniques of **Bag of Words (BoW)**, **Term Frequency-Inverse Document Frequency (TF-IDF)**, and **Unigrams/Bigrams**, which are often used to represent text data numerically.

#### **Basics Terminology used in NLP**

- **Corpus**: A large and structured set of texts that are used for training and testing language models. It can contain a wide variety of documents, such as books, articles, or web pages, and is used to help the model learn patterns in language. A corpus might be labeled (with annotations like sentiment or topics) or unlabeled.

- **Documents**: These are individual pieces of text in a corpus. A document can be anything from a sentence to a paragraph or an entire article, depending on the task at hand. In NLP, documents are often the basic units of analysis.

- **Vocabulary**: The set of unique words or tokens present in a corpus or document. It is the collection of all the words that the NLP model will learn from, and can vary in size depending on the corpus. For instance, a vocabulary for a specific task might include only words relevant to that task, like technical terms in a medical corpus.

- **Words**: Individual elements of a text, typically made up of one or more characters. Words are the basic building blocks in NLP and are often processed by tokenizing the text into smaller units. Words can be simple (like "cat") or complex (like "running" or "better").

![](https://miro.medium.com/v2/resize:fit:1100/format:webp/1*bwn2ItJqhBx9UYp6hkN7Cg.png)

#### 1. **Bag of Words (BoW)**

The **Bag of Words** model is one of the simplest and most common techniques for converting text into numerical data. It involves representing text as a set of words, ignoring grammar and word order but keeping track of the frequency of each word in the text.

##### Steps for BoW:
- Tokenize the text into individual words.
- Count the frequency of each word in the document.
- Create a vocabulary of unique words across the entire corpus.
- Represent each document as a vector, where each element corresponds to the frequency of a word from the vocabulary.

**Example**:

Text 1: `"I love programming"`
Text 2: `"I love Python programming"`

Vocabulary: `["I", "love", "programming", "Python"]`

Document 1 Vector: `[1, 1, 1, 0]`  
Document 2 Vector: `[1, 1, 1, 1]`

BoW is often implemented using tools like **CountVectorizer** in `scikit-learn`.

#### 2. **Term Frequency-Inverse Document Frequency (TF-IDF)**

TF-IDF is a more advanced technique that aims to reflect the importance of a word in a document relative to a collection of documents (corpus). It reduces the weight of frequently occurring words across all documents (like "the", "is", etc.) and increases the weight of words that are unique to a particular document.

##### Formula:
- **Term Frequency (TF)**: The number of times a word appears in a document divided by the total number of words in the document.
  - TF(t, d) = count of term t in document d / total number of terms in document d
  
- **Inverse Document Frequency (IDF)**: A measure of how much information the word provides, calculated as the logarithm of the inverse of the fraction of documents containing the word.
  - IDF(t, D) = log(total number of documents / number of documents containing term t)
  
- **TF-IDF**: The product of TF and IDF for each term in the document.
  - TF-IDF(t, d, D) = TF(t, d) * IDF(t, D)

**Example**:

If a word appears frequently in a document but not in others, its TF-IDF score will be higher, indicating that the word is important to the document. Words that are common across all documents will have lower TF-IDF scores.

TF-IDF is often implemented using **TfidfVectorizer** in `scikit-learn`.

#### 3. **Unigrams and Bigrams**

- **Unigrams**: These are single words or tokens. When text is tokenized, each individual word becomes a unigram.
  - Example: `"I love programming"` becomes `["I", "love", "programming"]`.

- **Bigrams**: These are pairs of consecutive words. When tokenizing text into bigrams, you capture more contextual information by considering word pairs.
  - Example: `"I love programming"` becomes `["I love", "love programming"]`.

In text preprocessing, unigrams and bigrams are used to capture the essence of the text at different granular levels:
- **Unigrams** focus on the individual importance of words.
- **Bigrams** capture the relationships between adjacent words, which may offer more context or meaning (e.g., "not good" vs. "good").

##### Example of Unigrams and Bigrams:
- Text: `"I love NLP programming"`
- Unigrams: `["I", "love", "NLP", "programming"]`
- Bigrams: `["I love", "love NLP", "NLP programming"]`

In practice, tools like **CountVectorizer** or **TfidfVectorizer** in `scikit-learn` allow for easy creation of both unigrams and bigrams by setting the `ngram_range` parameter:
- `ngram_range=(1,1)` for unigrams.
- `ngram_range=(1,2)` for both unigrams and bigrams.

#### Why Use BoW, TF-IDF, and N-grams?

- **Bag of Words** provides a simple and efficient way to represent text but lacks the ability to capture context or word order.
- **TF-IDF** is more advanced and useful in scenarios where word importance relative to the corpus matters, such as in document classification or information retrieval.
- **Unigrams and Bigrams** capture different levels of information, with bigrams offering more contextual insight, which can be useful for tasks like sentiment analysis or machine translation.

#### Conclusion:

Text preprocessing with techniques like **BoW**, **TF-IDF**, and **Unigrams/Bigrams** is crucial in converting raw text into a numerical format that machine learning models can understand. While BoW provides a straightforward approach, TF-IDF helps capture important terms, and unigrams/bigrams help represent both individual and contextual information in text.


#### **2.1 NLP | Text Preprocessing | One-Hot Encoding**

**One-Hot Encoding** is a technique used to convert categorical data, such as words, into a binary vector representation. Each word or token is represented as a vector where only the position corresponding to the word is marked with `1`, and all other positions are marked as `0`. This representation is useful in machine learning models to treat words as unique features.

![](https://miro.medium.com/v2/resize:fit:1100/format:webp/1*GsKLFAlzoNeIKo-1gz1h_Q.png)

### 1. **How One-Hot Encoding Works**

- **Vocabulary Creation**: First, a vocabulary (a list of unique words) is created from the entire dataset.
- **Vector Representation**: Each word is then represented as a binary vector. If the word is present in the document, the corresponding position in the vector is marked `1`, and all other positions are `0`.

For example, for the text:
- **Document 1**: "I love programming"
- **Document 2**: "I love machine learning"

The vocabulary is: `["I", "love", "programming", "machine", "learning"]`

The corresponding One-Hot encoded vectors would be:
- **Document 1**: `[1, 1, 1, 0, 0]`
- **Document 2**: `[1, 1, 0, 1, 1]`

Each vector corresponds to the presence or absence of a word in the vocabulary.


### 2. **Advantages of One-Hot Encoding**

- **Simple Representation**: One-hot encoding provides a straightforward way of representing text data in vector form.
- **No Ordering Assumptions**: Unlike techniques like **Bag of Words**, one-hot encoding does not assume any order of the words, which can be advantageous for certain models.
- **Easy to Implement**: One-hot encoding is easy to understand and implement using libraries like **scikit-learn**.


### 3. **Challenges of One-Hot Encoding**

- **High Dimensionality**: The size of the one-hot encoded vector grows with the size of the vocabulary. This can lead to very large vectors for datasets with large vocabularies.
- **Sparse Vectors**: One-hot vectors are often sparse, with many zeros in the vector. This sparsity can be inefficient in terms of memory usage.
- **No Semantic Information**: One-hot encoding does not capture any semantic meaning or relationships between words. For instance, "cat" and "dog" will be represented as independent, unrelated vectors, even though they are similar in meaning.


### 4. **Modern Alternatives**

- **TF-IDF (Term Frequency-Inverse Document Frequency)**: Unlike one-hot encoding, TF-IDF assigns weights to words based on their frequency in a document and their importance across the entire corpus.
- **Word Embeddings**: Methods like **Word2Vec** or **GloVe** create dense vector representations that capture semantic relationships between words, which one-hot encoding fails to do.


### 5. **Conclusion**

One-Hot Encoding is a simple and effective way to represent text data, particularly when the vocabulary is small. However, for large datasets or when semantic relationships between words are important, more advanced methods like **TF-IDF** or **Word Embeddings** are preferred.


#### **2.2 NLP | Text Preprocessing | BOW (Bag of Words)**

**Bag of Words (BOW)** is a simple and widely used method for converting text into a representation that can be used in machine learning models. The key idea is to treat each document as a collection of words (or tokens) without considering the order or structure of the words.

![](https://miro.medium.com/v2/resize:fit:1100/format:webp/1*axffCQ9ae0FHXxhuy66FbA.png)

### 1. **How BOW Works**

- **Tokenization**: Breaks the text into individual words or tokens.
- **Vocabulary Creation**: A list of all unique words in the text corpus is generated.
- **Vectorization**: Each document is represented as a vector of word counts, where the dimensions correspond to the words in the vocabulary.

For example, for the text:
- **Document 1**: "I love programming"
- **Document 2**: "I love machine learning"

The vocabulary is: `["I", "love", "programming", "machine", "learning"]`

The corresponding BOW vectors would be:
- **Document 1**: `[1, 1, 1, 0, 0]`
- **Document 2**: `[1, 1, 0, 1, 1]`

Each position in the vector corresponds to the count of the respective word in the document.


### 2. **Advantages of BOW**

- **Simplicity**: BOW is easy to implement and understand.
- **Effective for Text Classification**: It works well for many text classification tasks, like spam detection and sentiment analysis.
- **Flexibility**: It can handle any text data, including long documents or short messages.


### 3. **Challenges of BOW**

- **Loss of Context**: BOW ignores the order and structure of words, which means that context is lost. For example, "dog bites man" and "man bites dog" would be treated as the same.
- **High Dimensionality**: The vocabulary can grow very large, especially for large datasets, leading to high-dimensional feature vectors.
- **Sparse Representation**: Many words in the vocabulary may not appear in a particular document, leading to sparse vectors.


### 4. **Modern Alternatives**

- **TF-IDF (Term Frequency-Inverse Document Frequency)**: Addresses some of the limitations of BOW by giving more importance to less common words.
- **Word Embeddings**: Techniques like Word2Vec and GloVe create dense vector representations that capture semantic relationships between words.


### 5. Cosine Similarity

#### Definition  
A metric used to measure how similar two vectors are, based on the cosine of the angle between them.

![](https://cdn.botpenguin.com/assets/website/Screenshot_2024_03_13_at_1_44_43_PM_66fb6bcba4.webp)

#### Purpose  
Commonly used to compare text documents in NLP.

#### Formula  
```
Cosine Similarity = (A · B) / (||A|| ||B||)
```
Where:  
- A and B are vectors.  
- · represents the dot product.  
- ||A|| and ||B|| are the magnitudes of the vectors.

#### **2.4 NLP | Text Preprocessing | TF-IDF**

**TF-IDF (Term Frequency-Inverse Document Frequency)** is a numerical statistic used to reflect the importance of a word in a document relative to a collection or corpus of documents. It is widely used in text mining and information retrieval systems for tasks such as document classification, keyword extraction, and text summarization.

![](https://miro.medium.com/v2/resize:fit:1100/format:webp/1*XeYB3-C2cD3KIi975ql4_g.png)
![](https://miro.medium.com/v2/resize:fit:1100/format:webp/1*9tecGbVZZ8wxniSPoiXcYA.png)

### 1. **How TF-IDF Works**

TF-IDF combines two components:

- **Term Frequency (TF)**: Measures how frequently a word appears in a document. It is calculated as the number of times a word appears in a document divided by the total number of words in the document.
```
  TF(t) = (Number of times term t appears in a document) / (Total number of terms in the document)
```
- **Inverse Document Frequency (IDF)**: Measures how important a word is across the entire corpus. Words that appear in many documents have lower IDF scores, and words that appear in fewer documents have higher IDF scores. It is calculated as:
```
  IDF(t) = log(Total number of documents / Number of documents containing term t)
```
- **TF-IDF Calculation**: The final TF-IDF score for a term is the product of the TF and IDF values:
```
  TF-IDF(t) = TF(t) * IDF(t)
```
This score reflects both the frequency of a word in a specific document and how unique or important it is in the whole corpus.


### 2. **Advantages of TF-IDF**

- **Importance-based Weighting**: Unlike Bag of Words or One-Hot Encoding, TF-IDF gives more importance to words that are **unique** to specific documents, helping to capture important keywords and concepts.
- **Reduction of Common Words**: It reduces the weight of frequently occurring but less informative words like "the", "is", or "and" by assigning them low IDF values.
- **Effective for Information Retrieval**: TF-IDF is widely used in search engines and document retrieval systems, where it helps to rank documents based on their relevance to a query.


### 3. **Challenges of TF-IDF**

- **High Dimensionality**: Similar to One-Hot Encoding, TF-IDF creates a vector for each document, which can be very high-dimensional when working with large vocabularies.
- **No Context or Semantic Meaning**: While TF-IDF helps in identifying important terms, it still does not capture the semantic meaning of words or the context in which they are used.
- **Sensitive to Corpus Size**: TF-IDF works best when the corpus is relatively large and diverse. Small datasets may lead to skewed results.


### 4. **Modern Alternatives**

- **Word Embeddings**: Techniques like Word2Vec, GloVe, or FastText use dense vectors to represent words, capturing semantic relationships between words and reducing dimensionality.
- **Latent Semantic Analysis (LSA)**: LSA can be used to capture the underlying structure of the text, identifying topics and latent meanings in a document collection, further improving text representation.


### 5. **Conclusion**

TF-IDF is a powerful and widely-used technique for extracting important words from text data. It balances term frequency and document importance, making it suitable for tasks like document classification and search engine optimization. However, for capturing deeper semantic relationships, word embeddings or other modern techniques may be more appropriate.

### **3. NLP | Text Preprocessing | Word2Vec, Avg Word2Vec & Modern Techniques**

In text preprocessing for NLP, word embeddings are a crucial step in converting words into numerical representations. Embeddings capture semantic relationships between words, allowing machine learning models to understand textual data better. In this section, we will explore **Word2Vec**, **Average Word2Vec**, and modern techniques like **FastText** and **BERT**.

#### 1. **Word2Vec**

**Word2Vec** is a popular method for generating word embeddings that capture the semantic meaning of words based on their context in large text corpora. It was introduced by Google and is based on two types of models:
- **Continuous Bag of Words (CBOW)**: Predicts a target word from a fixed-size context window of surrounding words.
- **Skip-gram**: Predicts the surrounding context words given a target word.

![](https://miro.medium.com/v2/resize:fit:1100/format:webp/1*21Mvr4enfU78nstSZdP3Hg.png)
![](https://miro.medium.com/v2/resize:fit:828/format:webp/1*SFuHVAWXWB914rSGAZYLzA.png)
![](https://miro.medium.com/v2/resize:fit:1100/format:webp/1*3Eh6afALesGJ_1fkmBxxnQ.png)

##### How Word2Vec works:
- It uses a neural network model to map words into a continuous vector space.
- Words that appear in similar contexts are mapped to nearby points in the vector space, meaning they have similar representations.

**Example**:
- Word2Vec might map the words "king" and "queen" to vectors that are close in the vector space because they share similar relationships with other words like "man" and "woman."

Word2Vec can be implemented in Python using libraries like **Gensim**.

#### 2. **Average Word2Vec**

The **Average Word2Vec** approach is a simple yet effective way to represent entire sentences or documents. It involves averaging the word embeddings of all the words in the text.

##### Steps:
- For each word in the document, obtain its corresponding Word2Vec vector.
- Average the word vectors to create a single fixed-size vector for the entire document or sentence.
  
This method helps create a numerical representation of a document while maintaining some semantic meaning by utilizing individual word vectors.

**Example**:
For the sentence: `"I love NLP"`
- Word2Vec vectors for "I", "love", and "NLP" are obtained.
- These vectors are averaged to produce a single vector representing the sentence.

This approach is efficient but can lose some contextual information compared to more advanced techniques.

#### 3. **Modern Techniques for Word Embeddings**

While Word2Vec and Average Word2Vec are effective, modern NLP techniques have introduced more sophisticated ways of embedding words and entire sentences:

##### a. **FastText**

**FastText** is an extension of Word2Vec, developed by Facebook. Unlike Word2Vec, which treats each word as an indivisible unit, FastText breaks words into smaller units (subwords or character n-grams). This allows it to capture morphological information and handle rare or out-of-vocabulary words better.

**Advantages of FastText**:
- It can generate word embeddings for out-of-vocabulary words by leveraging subword information.
- Better at handling inflectional forms and compound words, as it represents words based on their subword units.

FastText is available in **Gensim** and **Facebook's FastText library**.

##### b. **BERT (Bidirectional Encoder Representations from Transformers)**

**BERT** is a transformer-based model developed by Google that provides context-aware word embeddings. Unlike Word2Vec and FastText, which generate a single static vector for each word, BERT generates dynamic word representations depending on the context of the word in a sentence.

- **Contextual embeddings**: BERT's embeddings vary depending on the surrounding words, meaning that the same word can have different embeddings in different contexts.
- **Transformer architecture**: BERT uses a bidirectional transformer model that reads the entire sentence at once, understanding context from both directions (left-to-right and right-to-left).

**Advantages of BERT**:
- It captures complex relationships between words by considering the full context of each word in the sentence.
- BERT embeddings are useful for a wide range of NLP tasks, including question answering, text classification, and named entity recognition (NER).

BERT and its variants (like **RoBERTa**, **DistilBERT**, and **ALBERT**) can be used through libraries like **Hugging Face's Transformers**.

##### c. **GPT (Generative Pre-trained Transformer)**

**GPT** (and its variants like **GPT-3** and **GPT-4**) is another modern language model by OpenAI. Unlike BERT, which is designed for bidirectional context, GPT uses a unidirectional transformer (left-to-right) and is optimized for text generation tasks.

**Advantages of GPT**:
- It's highly capable of generating human-like text and completing tasks such as translation, summarization, and conversation.
- GPT models can be fine-tuned for various specific tasks, including text classification and sentiment analysis.

#### Conclusion

- **Word2Vec** and **Average Word2Vec** remain widely used for creating word embeddings, providing a simple and efficient way to represent words and sentences.
- **FastText** improves on Word2Vec by leveraging subword information, making it more robust for rare and out-of-vocabulary words.
- **BERT** and its variants represent a major advancement in NLP, providing context-aware embeddings that are useful for a wide variety of tasks and outperform older models like Word2Vec in many benchmarks.
- **GPT** is another powerful model that excels at text generation and can be fine-tuned for different applications.

These techniques provide a range of options depending on the complexity of the task and the available resources.

##### **3.1.1 NLP | Text Preprocessing | Word Embedding | CBOW**

**Continuous Bag of Words (CBOW)** is one of the two main architectures used in the Word2Vec model for learning word embeddings. CBOW is designed to predict a target word based on its surrounding context words within a fixed-sized window.


### 1. **How CBOW Works**

In the CBOW architecture, the model takes a set of context words as input and tries to predict the target word. The context words are typically surrounding words within a predefined window size.

- **Input**: The context words (e.g., words surrounding the target word in a window).
- **Output**: The target word (the word we want to predict).

For example, for a sentence like "The cat sat on the mat," if the window size is 2, the context words for predicting the word "sat" would be "The," "cat," "on," and "the."

![](https://miro.medium.com/v2/resize:fit:1100/format:webp/0*i1Kxk51v-cig9cVI)

### 2. **Key Features of CBOW**

- **Contextual Information**: CBOW uses the surrounding context words to predict the target word. The more context words used, the more information the model has to learn word representations.
  
- **Word Vector Representations**: The model assigns a unique vector to each word in the vocabulary and learns to map similar words to nearby vectors in the vector space.

- **Window Size**: The window size is the number of context words considered on each side of the target word. A larger window size captures broader context, while a smaller window size focuses on more local relationships.


### 3. **CBOW Training Process**

[CBOW](https://www.youtube.com/watch?v=hQwFeIupNP0)

1. **Initialize Word Vectors**: Initially, the model assigns random vector representations to each word in the vocabulary.

>  - The vocabulary is unique words given by the corpus.
>  - no of neuron in hidden layer is randomly selected. it is the size of embedding vector where it is a hyperparameter

2. **Predict the Target Word**: For each training example (context, target), the model predicts the probability distribution over all possible words in the vocabulary.

3. **Optimization**: The model adjusts the word vectors to minimize the prediction error using techniques like stochastic gradient descent (SGD). The goal is to reduce the difference between the predicted word vector and the actual target word vector.

> ### Components of Neural Network in CBOW
>
> **1. Input Layer**  
> - Set of one-hot encoded vectors representing the context words (surrounding words of the target word).  
> - For a context size of 2m, the input layer will have 2m (m words on each side of the target word) one-hot vectors (one for each context word).  
>
> **2. Projection Layer (Hidden Layer)**  
> - A shared weight matrix maps the one-hot vectors into dense embeddings.  
> - Weight matrix dimensions: V x N, where:  
>   - V: Vocabulary size.  
>   - N: Embedding size.  
> - Computes the **average** of the context word embeddings.  
>
> **3. Output Layer**  
> - Applies a **softmax function** to produce a probability distribution over the vocabulary.  
> - Predicts the target word based on this probability.  
>
> **4. Loss Function**  
> - Minimizes the negative log-likelihood of predicting the correct target word given the context.  
> - Typically uses **cross-entropy loss** for training.

### 4. **Advantages of CBOW**

- **Efficient**: CBOW is relatively computationally efficient since it processes context words together and tries to predict a single target word.
  
- **Good for Frequent Words**: CBOW works well for learning embeddings for frequent words, as it focuses on context in a global manner.

- **Smooth Predictions**: The model learns embeddings by averaging the context words, resulting in smoother, more general representations of words.


### 5. **Disadvantages of CBOW**

- **Less Sensitive to Rare Words**: CBOW may not perform as well with rare or out-of-vocabulary words, as it relies on surrounding context.
  
- **Loss of Local Relationships**: Since it focuses on global context, CBOW may not capture some of the fine-grained local relationships between words that might be important in certain tasks.


### 6. **Applications of CBOW**

- **Word Similarity and Analogy**: CBOW embeddings can be used for tasks such as finding similar words or performing word analogies (e.g., "king" is to "queen" as "man" is to "woman").

- **Text Classification**: The word embeddings learned by CBOW can be used as features in text classification models to improve accuracy.

- **Sentiment Analysis**: CBOW embeddings help in sentiment analysis by providing rich semantic representations of words, capturing contextual relationships in phrases.


### 7. **Conclusion**

CBOW is an efficient and widely used method for learning word embeddings, particularly suitable for capturing the semantic relationships between words based on their surrounding context. Although it works well for learning global word relationships, it may not always capture finer nuances of meaning, especially for rare or specific words. Despite these limitations, CBOW remains a fundamental approach in many NLP applications and provides a foundation for understanding more advanced models like Skip-Gram, GloVe, and other embedding techniques.


##### **3.1.2 NLP | Text Preprocessing | Word Embedding | Skipgram**

**Skipgram** is another popular word embedding model in the Word2Vec family, developed by Mikolov et al. Skipgram is designed to predict the surrounding context words given a target word. This is the opposite of the Continuous Bag of Words (CBOW) model, which predicts the target word based on context words.


### 1. **How Skipgram Works**

In the Skipgram model, the input is a single word (the target word), and the output is a set of context words that surround it within a defined window size. The goal is to predict the probability of each context word given the target word.

For example, for a sentence like "The cat sat on the mat," and a window size of 2, if the target word is "sat," the context words would be "The," "cat," "on," and "the."

![](https://miro.medium.com/v2/resize:fit:1100/format:webp/0*4SSWryCfhS5mus2q)

image.png

### 2. **Key Features of Skipgram**

- **Target Word**: Skipgram uses a single target word as the input.
  
- **Context Words**: The model attempts to predict multiple context words surrounding the target word, based on a specified window size.

- **Prediction Task**: Skipgram's primary task is to predict context words for each target word, with the context being defined by a window around the target word.

- **Output**: The model outputs the probabilities for each word in the vocabulary to be a context word for the given target word.


### 3. **Skipgram Training Process**

1. **Initialize Word Vectors**: Similar to CBOW, Skipgram starts with random word vectors for all words in the vocabulary.

2. **Predict Context Words**: For each target word in the training set, Skipgram predicts the context words based on the surrounding window.

3. **Optimization**: Skipgram uses a technique like stochastic gradient descent (SGD) to optimize the model's word vectors by reducing the prediction error.

4. **Objective**: The objective is to maximize the probability of the actual context words given the target word.


### 4. **Advantages of Skipgram**

- **Good for Rare Words**: Skipgram performs well with infrequent words because it tries to predict the context words from the target word. Rare words tend to have a larger number of context words, which helps the model learn better representations for them.

- **Better Local Context Capture**: Skipgram tends to focus more on local context, making it effective for capturing relationships between words that appear together in smaller contexts.

- **Flexible Context Window**: Skipgram allows adjusting the window size, which can be optimized for different types of tasks (larger windows capture broader context, while smaller windows focus on more specific relationships).


### 5. **Disadvantages of Skipgram**

- **Slower for Large Corpora**: Since Skipgram uses a single word to predict multiple context words, it requires more computational power compared to CBOW, especially when dealing with large corpora.

- **Less Efficient with Common Words**: While Skipgram excels at learning embeddings for rare words, it may not perform as efficiently for frequent words, as these words are often repeated many times in the context of the corpus.

- **Training Time**: Skipgram may require more training time compared to CBOW, especially when the vocabulary size is large, as it needs to update weights for multiple context words for each target word.


### 6. **Applications of Skipgram**

- **Word Similarity**: Skipgram is widely used to find semantically similar words, as it captures relationships based on context words.
  
- **Word Analogy**: Skipgram can be used for tasks like word analogies (e.g., "man" is to "woman" as "king" is to "queen") by using vector arithmetic on the learned embeddings.

- **Information Retrieval**: Skipgram's embeddings can be used in information retrieval systems to better match query terms with relevant documents based on semantic similarity.

- **Text Classification**: Skipgram embeddings are often used as features in text classification tasks, where capturing semantic relationships is important.


### 7. **Conclusion**

Skipgram is an effective word embedding model, particularly for learning representations of rare words. It works by using a target word to predict the surrounding context words, and its ability to learn fine-grained relationships between words makes it useful for various NLP tasks. While Skipgram is computationally more intensive than CBOW and can be slower on larger corpora, it excels in capturing local word relationships and is particularly effective for tasks that require detailed semantic understanding, such as word analogy and word similarity.

#### **3.2 NLP | Text Preprocessing | Average Word2Vec**

**Average Word2Vec** is a technique used to represent entire sentences or documents as fixed-length vectors by averaging the Word2Vec embeddings of the words contained in them. Unlike individual word embeddings, which represent words in a continuous vector space, the Average Word2Vec method aggregates these word-level embeddings into a single vector that captures the overall meaning of a sentence or document.

![](https://i0.wp.com/yaronvazana.com/wp-content/uploads/2018/09/average-vectors.png?w=698&ssl=1)

### 1. **How Average Word2Vec Works**

In the Average Word2Vec model, each word in a sentence or document is represented by its corresponding Word2Vec embedding. These word embeddings are then averaged to create a single vector representation for the entire sentence or document. The resulting vector captures the semantic meaning of the text based on the individual word embeddings.

For example, consider the sentence: "The cat sat on the mat."

- The embeddings for each word ("The", "cat", "sat", "on", "the", "mat") are obtained using a pre-trained Word2Vec model.
- The embeddings are then averaged to produce a single vector that represents the entire sentence.


### 2. **Steps in Average Word2Vec**

1. **Obtain Word Embeddings**: For each word in the text, obtain its corresponding Word2Vec embedding vector. These embeddings can be pre-trained (e.g., using a model trained on a large corpus such as Google News) or trained on the specific corpus at hand.

2. **Vector Averaging**: Once all the word embeddings are retrieved, the next step is to calculate the mean (average) of all the word vectors in the sentence or document. This is done by summing up the vectors of all words and then dividing by the number of words.

3. **Generate Sentence/Document Representation**: The average vector serves as the representation of the entire sentence or document, capturing the semantic meaning based on the individual words.


### 3. **Advantages of Average Word2Vec**

- **Simplicity**: The method is straightforward to implement, making it easy to use in various NLP applications without requiring complex model architectures.

- **Fixed-Length Representation**: The resulting vector has a fixed length, regardless of the length of the input text. This makes it useful for tasks that require uniform vector inputs, such as machine learning models.

- **Good for Similarity Tasks**: Average Word2Vec representations can be used to measure the similarity between sentences or documents by calculating the cosine similarity between their respective vectors.

- **Handles Variable-Length Text**: Since the method aggregates information from each word in the text, it allows for the representation of variable-length sentences or documents with a fixed-length vector.


### 4. **Disadvantages of Average Word2Vec**

- **Loss of Word Order**: By averaging word embeddings, the method discards information about the order of words in the sentence. This can lead to a loss of important syntactic or grammatical relationships that affect the meaning.

- **Lack of Contextual Information**: Average Word2Vec does not take into account the context in which words appear, as it simply averages word vectors. This means that it may not capture nuances or polysemy (multiple meanings of words) effectively.

- **Insensitive to Word Importance**: All words are treated equally when averaging, which may not be ideal for tasks where some words (e.g., nouns or verbs) are more important than others.


### 5. **Applications of Average Word2Vec**

- **Document Classification**: Average Word2Vec vectors can be used as features for document classification tasks, where the goal is to categorize a document based on its overall content.

- **Sentiment Analysis**: This method can be used to represent entire sentences or reviews in sentiment analysis tasks, where the goal is to classify the sentiment as positive, negative, or neutral.

- **Semantic Similarity**: Average Word2Vec is commonly used to calculate the semantic similarity between sentences or documents. By comparing the average vectors of two texts, we can determine how similar they are in meaning.

- **Clustering and Topic Modeling**: Average Word2Vec representations are useful for grouping similar documents or sentences together in clustering or topic modeling tasks.


### 6. **Conclusion**

Average Word2Vec is a simple yet powerful method for generating fixed-length vector representations of sentences or documents. While it is efficient and works well for tasks like document classification and semantic similarity, it does have limitations, particularly in terms of preserving word order and context. For more complex NLP tasks that require deeper understanding of language structure, more advanced methods like transformer-based models (e.g., BERT, GPT) may be more suitable. However, Average Word2Vec remains a valuable technique for various applications that require efficient and scalable text representations.
