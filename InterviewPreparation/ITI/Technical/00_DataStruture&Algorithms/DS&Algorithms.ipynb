# 🚀 Phase 1: Foundations


## Big-O, Ω, Θ (Asymptotic Complexity Notations)

These notations are used in **algorithm analysis** to describe how runtime or space grows with input size `n`. They focus on *growth rate*, ignoring constants and lower-order terms.

---

### 1. Big-O Notation `O` — Upper Bound (Worst Case)

**Definition:**

f(n) = O(g(n)) if there exist constants c > 0 and n₀ such that:

f(n) ≤ c · g(n), for all n ≥ n₀

**Intuition:**
- Maximum growth rate
- Worst-case scenario
- Acts as a ceiling

**Example:**
- f(n) = 3n² + 5n + 2  
- O(n²)

---

### 2. Big-Omega Notation `Ω` — Lower Bound (Best Case)

**Definition:**

f(n) = Ω(g(n)) if there exist constants c > 0 and n₀ such that:

f(n) ≥ c · g(n), for all n ≥ n₀

**Intuition:**
- Minimum growth rate
- Best-case scenario
- Acts as a floor

**Example:**
- f(n) = 3n² + 5n + 2  
- Ω(n²)

---

### 3. Big-Theta Notation `Θ` — Tight Bound (Exact Growth)

**Definition:**

f(n) = Θ(g(n)) if there exist constants c₁, c₂ > 0 and n₀ such that:

c₁ · g(n) ≤ f(n) ≤ c₂ · g(n), for all n ≥ n₀

**Intuition:**
- Exact growth rate
- Both upper and lower bounds match

**Example:**
- f(n) = 3n² + 5n + 2  
- Θ(n²)

---

### Key Differences

| Notation | Meaning       | Bound Type   | Perspective   |
|----------|--------------|--------------|---------------|
| O        | Upper bound  | Worst case   | Max growth    |
| Ω        | Lower bound  | Best case    | Min growth    |
| Θ        | Tight bound  | Exact growth | Precise rate  |

---

### Examples

| Algorithm      | Best Case | Worst Case | Overall |
|---------------|----------|------------|---------|
| Linear Search | Ω(1)     | O(n)       | —       |
| Binary Search | Ω(1)     | O(log n)   | Θ(log n)|
| Merge Sort    | Ω(n log n)| O(n log n)| Θ(n log n)|

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None


def bst_search(root, target):
    current = root

    while current is not None:
        if target == current.value:
            return True
        elif target < current.value:
            current = current.left
        else:
            current = current.right

    return False
```

## 📌 Scientific Method for Calculating O, Ω, Θ (Asymptotic Complexity)

---

## 🧠 Core Principle

Any algorithm can be modeled as a cost function:

T(n) = sum of (cost of operation × frequency of execution)

Where each primitive operation has constant cost:

c = O(1)

So:

T(n) = Σ (c × f(n))

---

# 📌 STEP 1: Define Input Size

Identify the input variable(s):

- n → array size
- V, E → graph size
- n × n → matrix size

Everything must be expressed in terms of input growth.

---

# 📌 STEP 2: Identify Primitive Cost

Each basic operation is constant time:

- comparison
- assignment
- arithmetic operation
- memory access

So:

c = 1

---

# 📌 STEP 3: Translate Code Structure into Math

## Sequential execution:

T(n) = T1(n) + T2(n)

---

## Single loop:

T(n) = n × Tbody(n)

---

## Nested loop:

T(n) = n × (n × Tbody(n))

⇒ T(n) = n² × Tbody(n)

---

## k-level nested loop:

T(n) = nᵏ × Tbody(n)

---

# 📌 STEP 4: Worst Case (Big-O)

Definition:

T(n) ≤ c f(n), for all n ≥ n₀

### Method:
- choose input that maximizes execution
- expand full execution path
- keep dominant growth term

Result form:

T(n) ∈ O(f(n))

---

# 📌 STEP 5: Best Case (Big-Ω)

Definition:

T(n) ≥ c f(n), for all n ≥ n₀

### Method:
- choose input that minimizes execution
- remove worst branches
- keep minimal execution path

Result form:

T(n) ∈ Ω(f(n))

---

# 📌 STEP 6: Tight Bound (Big-Θ)

Definition:

T(n) ∈ Θ(f(n)) if:

T(n) ∈ O(f(n))  AND  T(n) ∈ Ω(f(n))

Meaning:

same asymptotic growth from both upper and lower bounds

---

# 📌 STEP 7: Dominant Term Rule

If:

T(n) = a n² + b n + c

Then:

- drop constants → a, b
- drop lower order terms → n, c

Result:

T(n) = Θ(n²)

---

# 📌 STEP 8: Recursion (General Case)

If recursion exists:

T(n) = a T(n / b) + f(n)

Solve using:

- Master Theorem
- Recursion Tree
- Substitution Method

---

# 📌 FINAL SCIENTIFIC WORKFLOW

To analyze ANY algorithm:

1. Define input size n
2. Identify primitive operations
3. Convert control flow → mathematical expression
4. Build T(n)
5. Derive:
   - Worst case → O(f(n))
   - Best case → Ω(f(n))
   - Tight bound → Θ(f(n))
6. Simplify using dominant term rule

---

# 🧠 KEY INSIGHT

You are analyzing growth:

lim (n → ∞) T(n)

Not code execution.

---

# 🚀 MENTAL MODEL

Algorithm → Cost Function → Asymptotic Growth Class

## 📌 Scientific Complexity Analysis (O, Ω, Θ) — BST Search

---

## 🧠 1. Problem Abstraction (Input Size)

Let:

- n = number of nodes in the BST
- h = height of the tree

Important:  
This algorithm does NOT scan all nodes.

It follows a single root-to-leaf path:

⇒ cost depends on h, not n

---

## 📌 2. Primitive Operation Cost Model

Each step performs constant-time operations:

- comparison: target == current.value
- comparison: target < current.value
- pointer movement: current = current.left / current.right

So:

c = 1

Each loop iteration costs O(1)

---

## 📌 3. Execution Structure Translation

The algorithm is a single loop:

T(n) = number of iterations × cost per iteration

So:

T(n) = k × 1

Where:

k = number of visited nodes along one path

⇒ k = h

Therefore:

T(n) = h

---

## 📌 4. Worst Case Analysis (Big-O)

### ⚠️ Worst-case input condition:

Tree is completely skewed:

1 → 2 → 3 → 4 → ... → n

So:

h = n

Substitute:

T(n) = n

### 📌 Result:

T(n) ∈ O(n)

---

## 📌 5. Best Case Analysis (Big-Ω)

### ⚡ Best-case input condition:

Target is found at root immediately.

So:

h = 1

Substitute:

T(n) = 1

### 📌 Result:

T(n) ∈ Ω(1)

---

## 📌 6. Balanced Structure Case (Typical Behavior)

### 🌳 Balanced BST property:

Each level reduces search space by half:

n → n / 2 → n / 4 → ... → 1

So:

h = log₂(n)

Substitute:

T(n) = log n

---

## 📌 7. Tight Bound Analysis (Big-Θ)

For a balanced BST:

- upper bound: O(log n)
- lower bound (typical growth): Ω(log n)

Therefore:

T(n) ∈ Θ(log n)

---

## 📌 8. Final Complexity Summary

| Case | Tree Structure | Height h | Complexity |
|------|---------------|----------|------------|
| Best Case | Root match | 1 | Ω(1) |
| Worst Case | Skewed list | n | O(n) |
| Average / Balanced | Balanced BST | log n | Θ(log n) |

---

## 🧠 Key Scientific Insight

The algorithm is fundamentally:

T(n) = O(h)

So complexity is determined by structural height:

- h = n  → linear growth
- h = log n → logarithmic growth

NOT by number of nodes directly.

---

## 🚀 Final Mental Model

Algorithm complexity is derived by:

1. Counting primitive operations per step
2. Mapping control flow → mathematical path length
3. Expressing runtime as:

T(n) = path length × O(1)
