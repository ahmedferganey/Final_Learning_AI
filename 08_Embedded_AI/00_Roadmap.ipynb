````python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime, timedelta

# Define phases with start and duration
phases = [
    ("Embedded Systems & Automotive Foundations", 0, 90),
    ("Computer Vision & Edge AI", 90, 90),
    ("Microservices & Cloud Integration", 180, 90),
    ("System Management & Monitoring", 270, 60),
    ("Final Projects & Portfolio Building", 330, 60)
]

# Define topics for each phase
topics = {
    "Embedded Systems & Automotive Foundations": [
        "Modern C++", "AUTOSAR", "CAN/LIN/FlexRay", "Zephyr RTOS", "AOSP/VHAL", "Git/GitHub"
    ],
    "Computer Vision & Edge AI": [
        "OpenCV", "MediaPipe", "YOLO/Ultralytics", "DeepSort", "PyTorch", "TFLite", "TensorRT",
        "Jetson Nano", "Raspberry Pi", "Model Quantization"
    ],
    "Microservices & Cloud Integration": [
        "FastAPI", "Flask", "PostgreSQL", "MongoDB", "RTSP/IP cameras", "Async I/O", 
        "Multithreading/Multiprocessing", "Docker", "MQTT", "Redis Streams", "AWS", "Azure"
    ],
    "System Management & Monitoring": [
        "MLflow", "Prometheus", "Grafana", "ELK Stack", "perf/gprof", "ADB/logcat", "GitHub Actions"
    ],
    "Final Projects & Portfolio Building": [
        "GitHub Portfolio", "Readme Writing", "LinkedIn/Blog", "Resume/CV"
    ]
}

# Gantt chart base
start_date = datetime(2025, 5, 1)
fig, ax = plt.subplots(figsize=(14, 8))

for i, (phase, offset, duration) in enumerate(phases):
    phase_start = start_date + timedelta(days=offset)
    phase_end = phase_start + timedelta(days=duration)
    ax.barh(phase, duration, left=mdates.date2num(phase_start), height=0.6, align='center', label=phase)

# Configure axis
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
ax.invert_yaxis()
plt.title("üìä Roadmap Gantt Chart: Path to Eagle Vision Role", fontsize=14)
plt.xlabel("Timeline")
plt.grid(True, which='both', axis='x', linestyle='--', linewidth=0.5)

plt.tight_layout()
plt.show()
````
