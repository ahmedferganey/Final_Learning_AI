# Hash Map

A **hash map** (or hash table) is a data structure that stores data in key-value pairs. It uses a hash function to compute an index into an array of buckets, allowing for efficient data retrieval.

## Key Features

1. **Key-Value Pairs**: 
   - Stores data as pairs, where each key is unique and maps to a specific value.

2. **Hash Function**:
   - Transforms the key into an index using a hashing algorithm. This index determines where the corresponding value is stored in the underlying array.

3. **Collision Handling**:
   - **Chaining**: Each slot contains a linked list (or another data structure) to handle multiple keys that hash to the same index.
   - **Open Addressing**: If a collision occurs, the hash map finds another open slot according to a probing sequence.

4. **Efficiency**:
   - Average-case time complexity for insertions, deletions, and lookups is O(1).
   - Worst-case time complexity can degrade to O(n) in scenarios with many collisions.

## Common Operations

- **Insertion**: Add a new key-value pair.
- **Lookup**: Retrieve the value associated with a specific key.
- **Deletion**: Remove a key-value pair from the map.

## Applications

- Used in scenarios requiring fast access to data, such as caching, databases, and associative arrays.


The **map()** function executes a specified function for each item in an iterable. The item is sent to the function as a parameter.

 - map(function, iterables)

Python **hash()** function is a built-in function and returns the hash value of an object if it has one. The hash value is an integer that is used to quickly compare dictionary keys while looking at a dictionary.

 - hash(obj)

````python
# Hash-MAP
class HashMap:
    def __init__(self, size=10):
        self.size = size
        self.map = [None] * self.size

    def _hash(self, key):
        # simple hash function
        return hash(key) % self.size
        # index = -4218233090008983961 % 10

    def set(self, key, value):
        index = self._hash(key)
        if self.map[index] is None:
            self.map[index] = []
    # checks whether the index calculated for a 
    # given key is currently empty (or unoccupied)
    # This condition checks if the slot at self.map[index]
    # is empty (i.e., None). If it is None, it means that there 
    # is currently no value stored at that index, indicating that 
    # it is available for insertion.    
        for i, (k, v) in enumerate(self.map[index]):
            """  
            This loop iterates over the list of tuples at self.map[index]. 
            It uses enumerate to get both the index (i) and the tuple ((k, v)) 
            containing the stored key (k) and value (v).
            """            
            if k == key:
                self.map[index][i] = (key, value)
                return 
        # If the key does not exist, append the new key-value pair
        self.map[index].append((key, value))
        """   
If the loop completes without finding the key, this line appends 
the new key-value pair as a tuple (key, value) to the list at self.map[index].        
        """
            
    def get(self, key):
        index = self._hash(key)
        if self.map[index] is not None:
            for k,v in self.map[index]:
                if k == key:
                    return v
        return None # key not found
    
    def remove(self, key):
        index = self._hash(key)
        if self.map[index] is not None:
            for i, (k, v) in enumerate(self.map[index]):
                if k == key:
                    del self.map[index][i]
                    return True
        return False
    
    def __str__(self):
        """Returns a user-friendly string representation of the hash map."""
        elements = []
        for index, bucket in enumerate(self.map):
            if bucket is not None:
                for k, v in bucket:
                    elements.append(f"{k}: {v}")
        return "{" + ", ".join(elements) + "}"
    

# Example usage:

# Initialize the hash map
hash_map = HashMap()
hash_map.set("Alice", {"name": "Alice", "age": 30})

# Retrieve Alice's data and access specific properties
alice_data = hash_map.get("Alice")
print(alice_data)           # Output: {'name': 'Alice', 'age': 30}
print(alice_data["age"])     # Output: 30
print(alice_data["name"])    # Output: Alice

# Remove 'age' from Alice's dictionary directly
if "age" in alice_data:
    del alice_data["age"]

# Update the hash map with the modified Alice data
hash_map.set("Alice", alice_data)

# Verify that 'age' was removed
print(hash_map.get("Alice")) # Output: {'name': 'Alice'}
print(hash_map)              # Output: {Alice: {'name': 'Alice'}}
````

**Problem:** Find the length of the longest substring without repeatingcharacters.

````python
def length_of_longest_substring(s):
    char_set = set()
    left = max_length = 0
    for right in range(len(s)):
        while s[right] in char_set:
            char_set.remove(s[left])
            left +=1
        char_set.add(s[right])
        max_length = max(max_length, right - left + 1)
    return max_length
    

s1 = "ajdnwgdcasfnxerivuhrlgcs"
s2 = "sddqwsahfgdashytwrgsxwwrkl"
s3 = "wqkfscwefqwmvrmipecvs"
s4 = "dkgmbywotnvbsgto"
print(length_of_longest_substring(s1))
print(length_of_longest_substring(s2))
print(length_of_longest_substring(s3))
print(length_of_longest_substring(s4))
````

**3. Median of Two Sorted Arrays**

**Problem:** Find the median of two sorted arrays.

````python
def find_median_sorted_arrays(nums1, nums2):
    merged = sorted(nums1+nums2)
    n = len(merged)
    if n % 2 == 0:
        return (merged[n // 2 -1] + merged[n // 2]) / 2 
    else:
        return merged[n // 2]

nums1 = [1,3,5,7,9]
nums2 = [12,13,22,45,54,89]

print(find_median_sorted_arrays(nums1, nums2))
````

**4. Longest Palindromic Substring**

**Problem:** Given a string, find the longest palindromic substring.

````python
def longest_palindrome(s):
    def expand_around_center(left, right):
        while left >= 0 and right < len(s) and s[left] == s[right]:
            left  -= 1
            right += 1
        return right - left - 1

    start = end = 0 # "babad"
    for i in range(len(s)):
        len1 = expand_around_center(i, i) # for the center
        len2 = expand_around_center(i, i+1) # if center between i, i+1 
        max_len = max(len1, len2)
        if max_len > end - start:
            start = i - (max_len - 1) // 2
            end = i + max_len // 2
````

# Longest Palindrome Algorithm Illustration

## Algorithm Overview
The algorithm finds the longest palindromic substring within a given string using the "expand around center" technique. It iterates through each character (and each pair of characters) in the string and checks for palindromes by expanding outward.

## Input
**String**: `s = "babab"`

## Initial Setup
- Initialize `start` and `end` to `0`: These will track the start and end indices of the longest palindrome found.

## Iteration Process

1. **Iteration 0 (i = 0, character = 'b')**
   - **Odd-length check**: Call `expand_around_center(0, 0)`
     - **Left**: 0, **Right**: 0 (characters: 'b' == 'b')
     - Expand: Left = -1, Right = 1 (out of bounds)
     - **Length**: `1` (return `1`)
   - **Even-length check**: Call `expand_around_center(0, 1)`
     - **Left**: 0, **Right**: 1 (characters: 'b' != 'a')
     - **Length**: `0` (return `0`)
   - **Max length**: `max(1, 0) = 1`
   - Update `start` and `end`: `start = 0`, `end = 0` (longest palindrome is "b").

2. **Iteration 1 (i = 1, character = 'a')**
   - **Odd-length check**: Call `expand_around_center(1, 1)`
     - **Left**: 1, **Right**: 1 (characters: 'a' == 'a')
     - Expand: Left = 0, Right = 2 (characters: 'b' == 'b')
     - Expand: Left = -1, Right = 3 (out of bounds)
     - **Length**: `3` (return `3`)
   - **Even-length check**: Call `expand_around_center(1, 2)`
     - **Left**: 1, **Right**: 2 (characters: 'a' != 'b')
     - **Length**: `0` (return `0`)
   - **Max length**: `max(3, 0) = 3`
   - Update `start` and `end`: `start = 1 - (3 - 1) // 2 = 0`, `end = 1 + 3 // 2 = 2` (longest palindrome is "bab").

3. **Iteration 2 (i = 2, character = 'b')**
   - **Odd-length check**: Call `expand_around_center(2, 2)`
     - **Left**: 2, **Right**: 2 (characters: 'b' == 'b')
     - Expand: Left = 1, Right = 3 (characters: 'a' == 'a')
     - Expand: Left = 0, Right = 4 (characters: 'b' == 'b')
     - Expand: Left = -1, Right = 5 (out of bounds)
     - **Length**: `5` (return `5`)
   - **Even-length check**: Call `expand_around_center(2, 3)`
     - **Left**: 2, **Right**: 3 (characters: 'b' != 'a')
     - **Length**: `0` (return `0`)
   - **Max length**: `max(5, 0) = 5`
   - Update `start` and `end`: `start = 2 - (5 - 1) // 2 = 0`, `end = 2 + 5 // 2 = 4` (longest palindrome is "bab").

4. **Iteration 3 (i = 3, character = 'a')**
   - **Odd-length check**: Call `expand_around_center(3, 3)`
     - **Left**: 3, **Right**: 3 (characters: 'a' == 'a')
     - Expand: Left = 2, Right = 4 (characters: 'b' == 'b')
     - Expand: Left = 1, Right = 5 (out of bounds)
     - **Length**: `3` (return `3`)
   - **Even-length check**: Call `expand_around_center(3, 4)`
     - **Left**: 3, **Right**: 4 (characters: 'a' != 'd')
     - **Length**: `0` (return `0`)
   - **Max length**: `max(3, 0) = 3`
   - No update needed as `max_len` is not greater than `end - start`.

5. **Iteration 4 (i = 4, character = 'd')**
   - **Odd-length check**: Call `expand_around_center(4, 4)`

   - No update needed as `max_len` is not greater than `end - start`.

# Detailed Breakdown of Iteration 1

## Iteration 1 Overview
In this iteration, we check for the longest palindrome centered at the character **'a'** (index **1**).

### Initial State
- **Current Character**: 'a' (at index 1)
- **Function Call**: `expand_around_center(1, 1)` for the odd-length palindrome.
- **Left Pointer**: 1 (index of 'a')
- **Right Pointer**: 1 (same index for odd-length palindromes)

### Step-by-Step Expansion

1. **First Expansion**:
   - Check if `s[left]` equals `s[right]`:
     - `s[1]` (which is 'a') equals `s[1]` (which is also 'a').
     - They are equal, so we expand:
       - **Decrement `left`**: 0
       - **Increment `right`**: 2
   - **New Pointers**: **Left = 0**, **Right = 2**

2. **Second Expansion**:
   - Check again if `s[left]` equals `s[right]`:
     - `s[0]` (which is 'b') equals `s[2]` (which is also 'b').
     - They are equal, so we expand again:
       - **Decrement `left`**: -1
       - **Increment `right`**: 3
   - **New Pointers**: **Left = -1**, **Right = 3**

### Termination Condition
Now we check the conditions to continue expanding:
- **Condition 1**: `left >= 0`
  - This condition **fails** because `left` is now **-1**.
- **Condition 2**: `right < len(s)`
  - This condition **succeeds** because `right` is **3**, and `len(s)` is **5**.
- **Condition 3**: `s[left] == s[right]`
  - This condition is not evaluated because we already failed the first condition (`left >= 0`).

Since the first condition fails (`left < 0`), we can no longer expand. This indicates we have reached the left boundary of the string.

### Length Calculation
The length of the palindrome found during the expansions is calculated as:
\[
\text{Length} = \text{right} - \text{left} - 1
\]
Substituting in the values:
- **Right**: 3
- **Left**: -1

The length calculation:
\[
\text{Length} = 3 - (-1) - 1 = 3
\]

Thus, the algorithm determines that the longest palindromic substring centered at index 1 has a length of **3**. The substring found is **"bab"**.

## Summary
The `Left = -1` and `Right = 3` indicates that the algorithm has successfully expanded beyond the bounds of the string to find the longest palindrome centered at the character 'a' (index 1). The expansion stops when one pointer goes out of bounds, which is a standard condition to prevent further unnecessary checks.


**5. Container With Most Water**

**Problem:** Given n non-negative integers representing the height of bars, find two lines that together with the x-axis form a container that holds the most water.

````python
def max_area(height):
    left, right = 0, len(height) -1 
    max_area = 0
    while left < right:
        width = right - left
        max_area = max(max_area, min(height[left], height[right]) * width)
        if height[left] < height [right]:
            left +=1
        else:
            right -=1
    return max_area


height = [1,2,4,1,57,4,2,1,32,3,61,2]
print(max_area(height))
````

**6. 3 Sum**

**Problem:** Given an integer array, return all the triplets [nums[i], nums[j], nums[k]] such that i != j, i != k, and j != k, and nums[i] + nums[j] + nums[k] == 0.

````python
def three_sum(nums):
    nums.sort()
    result = []
    n = len(nums)

    for i in range(n):
        if i > 0 and nums[i] == nums[i - 1]:
            #  to skip duplicate values in the sorted 
            #   array while iterating through the list. 
            continue # to turn to next iteration before implelemting the rest of loop 
        left, right = i+1, n - 1     # Set two pointers
        while left < right:
            total = nums[i] + nums[left] + nums[right]
            if total < 0:
                left +=1    # Move the left pointer to the right to increase the sum
            elif total > 0:
                right -=1   # Move the right pointer to the left to decrease the sum
            else:
                result.append([nums[i] , nums[left] , nums[right]])
                # Skip duplicates for the second and third numbers
                while left < right and nums[left] == nums[left + 1]:
                    left +=1 
                while left < right and nums[right] == nums[right - 1]:
                    right -=1
                left +=1
                right -=1
    return result

nums = [-1, 0, 1, 2, -1, -4]
print(three_sum(nums))
    
````

**7. Valid Parentheses**

**Problem:** Given a string containing just the characters (, ), {, }, [ and ], determine if the input string is valid.

````python
def is_valid(s): # "())([]"
    stack = []  # Initialize an empty stack to keep track of opening brackets.
    """
The stack is used to keep track of the opening brackets 
as they appear in the string. The main idea is to ensure 
that every closing bracket has a corresponding opening 
bracket and that they are correctly nested.
    """    
    mapping = {')': '(', '}': '{', ']': '['}  # Mapping of closing brackets to opening brackets.
    """  
Keys: The keys in this dictionary are the closing brackets: ')', '}', and ']'.
Values: The values are the corresponding opening brackets: '(', '{', and '['.    
    """
    for char in s:  # Iterate over each character in the input string.
        if char in mapping:  # Check if the character is a closing bracket.
            top_element = stack.pop() if stack else '#'  # Pop the top element from the stack or set a dummy value if the stack is empty.
            if mapping[char] != top_element:  # Compare the popped element with the corresponding opening bracket.
                return False  # If they don't match, the string is not valid.
        else:
            if char in mapping.values():  # If it's an opening bracket, push it onto the stack.
                stack.append(char)
    return not stack  # Return True if the stack is empty (valid string), False otherwise.


print(is_valid("()"))            # True
print(is_valid("()[]{}"))        # True
print(is_valid("(]"))            # False
print(is_valid("([)]"))          # False
print(is_valid("{[]}"))          # True
print(is_valid("asd()()[]"))          # True

print(is_valid("asd())[]")) # false
````

**Problem:** Given an integer array nums, return an array output such that output[i] is equal to the product of all the elements of nums except nums[i].

````python
def product_except_self(nums):
    length = len(nums)
    output = [1] * length  # Initialize the output array

    left_product = 1
    for i in range(length):
        output[i] = left_product    # output[i] will contain the product of all elements to the left
        left_product *=  nums[i] # Update left_product for the next index

    right_product = 1
    for i in range(length -1, -1, -1):
        output[i] *= right_product
        right_product *= nums[i]  
    return output      

# Example usage
nums = [1, 2, 3, 4]
print(product_except_self(nums))  # Output: [24, 12, 8, 6]
````

**13. Search in Rotated Sorted Array**

**Problem:** Suppose an array sorted in ascending order is rotated at some pivot unknown to you beforehand. You are given a target value to search. If found, return its index; otherwise, return -1.


````python
def search_in_rotated_sorted_array(nums, target):
    left, right =0, len(nums) -1
    while left <= right:
        mid = (left+right) // 2
        if nums[mid] == target:
            return mid
        if nums[left]  <= nums[mid]: # left half is sorted
            if nums[left] <= target < nums[mid]:
                right = mid-1
            else:
                left = mid +1
        else : # right half is sorted
            if nums[mid] < target <= nums[right]:
                left = mid +1
            else:
                right = mid -1
    return -1

index = search_in_rotated_sorted_array([4, 5, 6, 7, 0, 1, 2], 0)
index
````

### Key Requirements and Concepts

1. **Rotated Sorted Array**:
   - A rotated sorted array is an array that has been sorted in ascending order but then rotated at a certain pivot. For example, the sorted array `[0, 1, 2, 4, 5, 6, 7]` can be rotated to form `[4, 5, 6, 7, 0, 1, 2]`.
   - The pivot is the point of rotation; elements to the left of the pivot are greater than those to the right.

2. **Binary Search**:
   - Binary search is a classic algorithm for finding an element in a sorted array in \(O(\log n)\) time complexity. It works by repeatedly dividing the search interval in half.
   - In a rotated sorted array, we can still apply binary search principles with some modifications.

3. **Target Value**:
   - You are given a target value that you need to find in the rotated sorted array. Your goal is to return the index of this target if it exists in the array; otherwise, return -1.

### Understanding the Problem with an Example

#### Example Array

    [4, 5, 6, 7, 0, 1, 2]


- **Original Sorted Array**: If we consider the original sorted array before rotation, it would be `[0, 1, 2, 4, 5, 6, 7]`.
- **Target**: Suppose we want to find the target value `0`.

#### Steps to Find the Target

1. **Identify the Rotation**: In the given array, `4` is the largest element, followed by a smaller element `0`, indicating a pivot point.

2. **Binary Search Mechanism**:
   - Start with two pointers, `left` at index `0` (value `4`) and `right` at index `6` (value `2`).
   - Calculate the mid index: `mid = (left + right) // 2`.
   - Check if the `mid` value equals the target. If it does, return the index.

3. **Determine Which Side is Sorted**:
   - If the left part of the array (from `left` to `mid`) is sorted (i.e., `nums[left] <= nums[mid]`), check if the target lies within that range:
     - If it does, adjust the `right` pointer to search in the left half.
     - If it doesn’t, adjust the `left` pointer to search in the right half.
   - If the right part is sorted, do the opposite: check if the target is within that range and adjust the pointers accordingly.

4. **Repeat**: Continue this process until the target is found or the pointers converge, indicating the target is not in the array.

### Key Takeaways

- **Understanding the Structure**: Knowing how the rotation affects the order of elements is crucial for applying binary search effectively.
- **Using Binary Search**: The approach leverages the sorted properties of the subarrays formed by the rotation.
- **Edge Cases**: Consider edge cases such as arrays with only one element, arrays that are not rotated at all, or when the target is the smallest or largest element.

````python
from collections import defaultdict, deque

def can_finish(nums_courses, prerequisites):
    graph = defaultdict(list)
    indegree = [0] * nums_courses

    for a, b in prerequisites: # {[a,b], [c, d]} to take a must complete b
        graph[b].append(a)  # a depend on b
        indegree[a] += 1 # no of dependant
    
    queue = deque([i for i in range(nums_courses) if indegree[i] == 0])
    count = 0

    while queue:
        course = queue.popleft()
        count += 1
        for neighbor in graph[course]:
            indegree[neighbor] -=1
            if indegree[neighbor] == 0:
                queue.append(neighbor)
    return count == nums_courses
    
# Test case
num_courses = 4
prerequisites = [[0, 1], [0, 2], [1, 3]]  # Example where all courses can be completed

# Running the function and printing the output
can_finish(num_courses, prerequisites)    
````

**15. Maximum Subarray**

**Problem:** Find the contiguous subarray (containing at least one number) which has the largest sum and return its sum.

````python
def max_sub_array(nums):
    max_sum = current_sum = nums[0]
    for num in nums[1:]:
        current_sum = max(num, current_sum+num)
        max_sum = max(max_sum, current_sum)
    return max_sum
max_sub_array([0,1,-2,3,1,-3])
````

### Kadane's Algorithm for Maximum Subarray Sum

To solve the problem of finding the maximum sum of a contiguous subarray, we can use **Kadane's Algorithm**. This algorithm efficiently finds the maximum subarray sum in linear time.

#### Steps of Kadane's Algorithm:

1. **Initialize two variables:**
   - `max_current` to store the maximum sum of the subarray that ends at the current position.
   - `max_global` to store the overall maximum sum found so far.

2. **Loop through each element in the array:**
   - Update `max_current` to be the maximum of the current element itself or the sum of `max_current` and the current element. This step decides whether to add the current element to the existing subarray or start a new subarray with the current element.
   - Update `max_global` if `max_current` is greater than `max_global`.

3. **Return `max_global` as the result.**



### Example

Let's consider the following array:

```plaintext
nums = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
```

Steps of Kadane's Algorithm:

**Initialization:**

Set max_current and max_global to the first element of the array:

max_current = -2
max_global = -2

**Iterate through the array:**

 - For num = 1:

max_current = max(1, -2 + 1) = 1
max_global = max(-2, 1) = 1

 - For num = -3:

max_current = max(-3, 1 - 3) = -2
max_global = max(1, -2) = 1

 - For num = 4:

max_current = max(4, -2 + 4) = 4
max_global = max(1, 4) = 4

 - For num = -1:

max_current = max(-1, 4 - 1) = 3
max_global = max(4, 3) = 4

 - For num = 2:

max_current = max(2, 3 + 2) = 5
max_global = max(4, 5) = 5

 - For num = 1:

max_current = max(1, 5 + 1) = 6
max_global = max(5, 6) = 6

 - For num = -5:

max_current = max(-5, 6 - 5) = 1
max_global = max(6, 1) = 6

 - For num = 4:

max_current = max(4, 1 + 4) = 5
max_global = max(6, 5) = 6

**Final Result:**
After processing all elements, the maximum subarray sum is stored in max_global, which is 6.

**Subarray**
The subarray that gives this maximum sum is [4, -1, 2, 1], which has a sum of 6.

**Conclusion**
Using Kadane's Algorithm, we efficiently found the maximum sum of a contiguous subarray in linear time, resulting in a maximum sum of 6 for the given input array.

**16. Minimum Path Sum**

**Problem:** Given a grid filled with non-negative integers, find a path from the top left to the bottom right, which minimizes the sum of the numbers along its path.

````python
def min_path_sum(grid):
    if not grid:
        return 0
    for i in range(len(grid)):
        for j in range(len(grid[0])):
            if i == 0 and j == 0: # Check if we are at the top-left cell
                continue         # Skip any calculations for this cell
            top  = grid[i-1][j] if i > 0 else float('inf')
            """   
Purpose: This line determines the cost of moving from the cell 
directly above the current cell (i, j), which is (i-1, j).
            """            
            left = grid[i][j-1] if j > 0 else float('inf')
            """  
Purpose: This line determines the cost of moving from the 
cell directly to the left of the current cell (i, j), which is (i, j-1).            
            """            
            grid[i][j] += min(top, left)
    return grid[-1][-1]

CityGrid = [[1,2,3],
            [4,2,6],
            [2,3,2]]
min_path_sum(CityGrid)
````

### Approach

To solve this problem, we can use dynamic programming. The idea is to iterate through each cell in the grid and calculate the minimum path sum to reach that cell from the top left corner. We will create a 2D array `dp` where `dp[i][j]` represents the minimum path sum to reach the cell `(i, j)`.

### Steps

1. **Initialize the DP Table:**
   - Create a 2D array `dp` of the same dimensions as the grid.
   - Set `dp[0][0]` to `grid[0][0]` since the starting point has no preceding cells.

2. **Fill the First Row and First Column:**
   - For the first row, each cell can only be reached from the cell to its left.
   - For the first column, each cell can only be reached from the cell above it.

3. **Fill the Rest of the DP Table:**
   - For each cell `(i, j)`, calculate the minimum path sum using the formula:
     ```python
     dp[i][j] = grid[i][j] + min(dp[i-1][j], dp[i][j-1])
     ```
   - This formula takes the current cell's value and adds it to the minimum of the path sums from the cell above or the cell to the left.

4. **Return the Result:**
   - The minimum path sum to reach the bottom right corner will be stored in `dp[rows-1][cols-1]`.

````python
def unique_paths(m,n):
    dp = [[1]*n for _ in range(m)]
    for i in range(1, m):
        for j in range(1, n):
            dp[i][j] = dp[i-1][j]+dp[i][j-1]
    return dp[-1][-1]


"""  
# dp[i][j] = dp[i - 1][j] + dp[i][j - 1]
To determine the number of unique paths to cell (i, j), 
you need to consider how you could have arrived at that cell:

    From Above: You can come from the cell directly above it, 
    which is (i - 1, j).

    From the Left: You can also come from the cell directly 
    to the left of it, which is (i, j - 1).
"""
````

````python
from collections import Counter

# s = "cbaebabacd" and p = "abc"
def find_anagrams(s, p):
    p_count = Counter(p) # {'a': 1, 'b': 1, 'c': 1}
    s_count = Counter()
    result = []

    # sliding window technique 
    for i in range(s):
        """  
            At index 1 (character 'b'):
            s_count becomes {'c': 1, 'b': 1}        
        """
        s_count[s[i]] +=1 # s_count['c'] += 1  # s_count becomes Counter({'c': 1})
        if i >= len(p):
            """  
The sliding window is being used to analyze segments of s 
that have the same length as p. When the index i reaches or 
exceeds the length of p, the current window is considered 
"full," and we need to maintain its size by removing the 
character that is no longer in the window (the character 
at index i - len(p)).            
            """
            if s_count[s[i-len(p)]] ==1:
                del s_count[s[i-len(p)]]
            else:
                s_count[s[i-len(p)]] -=1
        if s_count == p_count:
            result.append(i- len(p) +1) 
            # after iterations 0,1,2 --> result.append(2 - 3 + 1)  # result becomes [0]

    return result
            
````

**20. Longest Increasing Subsequence**

**Problem:** Given an integer array, return the length of the longest increasing subsequence.

````python
import random
from time import time

# Generate a large list of random integers
large_data = [random.randint(1, 10**6) for _ in range(10**4)]  # Adjust 10**4 as needed for testing
````

````python
def length_of_lis(nums):
    if not nums:
        return 0
    dp = [1] * len(nums)
    """  
The dp array is initialized with the value 1 for each 
element because the minimum length of the increasing 
subsequence that includes each individual number is 
1 (the number itself).
    """
    for i in range(1, len(nums)):
        """  
The index i represents the current element for which 
we are trying to determine the length of the longest 
increasing subsequence that ends at nums[i].        
        """
        # For each i, this loop iterates from 0 to i-1 (all previous elements).
        for j in range(i): # The index j represents a previous element that we are comparing against nums[i].
            if nums[i] > nums[j]:
                dp[i] = max(dp[i], dp[j]+1)
                """  
 adding 1 to account for the inclusion of the current element 
 (nums[i]) in the increasing subsequence that ends at nums[j]                
                """
    return max(dp)

````

````python
import bisect
"""  
The bisect module in Python is used for maintaining a list 
in sorted order without having to sort the list after each 
insertion. It provides functions for performing binary search 
operations on sorted lists, allowing for efficient insertion 
and searching.
"""
def length_of_lis_binary_search(nums):
    if not nums:
        return 0
    lis = []
    """ 
lis = []: This list will store the smallest possible tail 
values of increasing subsequences. The idea is that for 
any increasing subsequence, the last element will be 
the smallest possible.    
    """

    for num in nums:
        pos = bisect.bisect_left(lis, num) 
        """ 
pos = bisect.bisect_left(lis, num): 
This line uses the bisect_left function 
to find the position pos in the lis where num can be inserted to maintain
the sorted order. Here's how it works:
        If num is less than the smallest element in lis, pos will be 0.
        If num is greater than all elements in lis, pos will be equal to the length of lis.
        If num is equal to an element in lis, pos will be the index of that element.        
        """
        if pos == len(lis):
            lis.append(num)
        else:
            lis[pos] = num
    return len(lis)
````

**1. Grid Path with Obstacles**

**Problem:** Given a grid where some cells have obstacles, find the number of unique paths from the top-left to the bottom-right, only moving right or down.
Application: This can simulate avoiding obstacles in an environment.



````python
"""  
In this problem, you need to calculate the number of unique paths 
from the top-left to the bottom-right of a grid that contains obstacles.
"""

def unique_paths_with_obstacles(grid):
    rows, cols = len(grid), len(grid[0])
    if grid[0][0] ==1 or grid[rows-1][cols-1] ==1:
        return 0 # no path if start or end is an obstacle

    dp = [[0] * cols for _ in range(rows)]
    """  
setting dp[0][0] = 1 establishes the initial condition: 
there's exactly one way to reach the starting cell if it 
is not an obstacle.
    """  
    dp[0][0] =1

    for i in range(rows):
        for j in range(cols):
            if grid[i][j] == 1:
                dp[i][j] = 0 # No path through obstacles
            else:
                """ 
Dynamic Programming Principle: This check ensures that the path counting 
is done correctly without attempting to access invalid indices.                
                """
                if i > 0:# Check if not in the first row
                    dp[i][j] += dp[i-1][j]# Paths from above
                if j > 0:# Check if not in the first column
                    dp[i][j] += dp[i][j-1]# Paths from the left
    return dp[-1][-1]

# Example usage:
grid = [
    [0, 0, 0],
    [0, 1, 0],
    [0, 0, 0]
]
print(unique_paths_with_obstacles(grid))  # Output: 2                    
````

**2. Minimum Fuel Cost in a Grid**

**Problem:** In a grid where each cell represents fuel cost, find the minimum fuel required to move from the top-left to the bottom-right.

**Application:** Helps in optimizing fuel usage while navigating terrain with varying fuel costs.



````python
def min_fuel_cost(grid):
    rows, cols = len(grid), len(grid[0])
    dp = [[float('inf')] * cols for _ in range(rows)]
    dp[0][0] = grid[0][0]

    for i in range(rows):
        for j in range(cols):
            if i > 0:
                dp[i][j] = min(dp[i][j], dp[i-1][j]+grid[i][j])
            if j > 0:
                dp[i][j] = min(dp[i][j], dp[i][j-1]+grid[i][j])
    return dp[-1][-1]

# Example usage:
fuel_grid = [
    [1, 2, 3],
    [4, 2, 6],
    [2, 3, 2]
]
print(min_fuel_cost(fuel_grid))  # Output: 10
````


**Optimal Sensor Placement**

**Problem:** Place sensors in a grid to maximize the area coverage with minimum sensors.

**Application:** Used in sensor placement for maximum coverage of an autonomous vehicle's surroundings.


````python
# using a greedy algorithm to place sensors in a grid. The coverage for 
# each sensor is assumed to be a simple square region around its placement.
"""  
                        Problem Breakdown

Grid Definition: We'll represent the area as a grid where each cell can 
either be covered by a sensor or not.

Coverage Range: Each sensor has a specific range within which it can 
detect its surroundings. This range defines how many adjacent cells 
can be considered "covered" by that sensor.

Objective: The goal is to maximize the total number of covered cells 
using the fewest sensors possible.
"""

""" 
To tackle the problem of Optimal Sensor Placement in a grid, we want 
to strategically place a limited number of sensors in such a way that
the maximum area is covered. This is particularly relevant for 
autonomous vehicles, where sensor placement can significantly affect 
the vehicle's ability to perceive its surroundings.
"""

import numpy as np 
import matplotlib.pyplot as plt

def create_grid(size):
    return np.zeros((size, size))

def place_sensor(grid, x, y, coverage_range):
    """Place a sensor on the grid and cover the specified area."""
    size = grid.shape[0]
    """  
The function uses nested loops to iterate over a square area 
centered around the sensor's position (x, y). The loops run 
from x - coverage_range to x + coverage_range 
and 
from y - coverage_range to y + coverage_range.    
    """
    for i in range(max(0, x-coverage_range), min(size, x + coverage_range+1)):
        for j in range (max(0, y-coverage_range) , min(size, y + coverage_range +1)):
            grid[i][j]  = 1 # mark covered area


def optimal_sensor_placement(grid_size, sensor_count, coverage_range):
    grid = create_grid(grid_size)
    sensors_placed = []

    for _ in range(sensor_count):
        best_position = None
        max_covered = -1

        # Try to place a sensor in each cell and calculate coverage
        for i in range(grid_size):
            for j in range(grid_size):
                if grid[i][j] == 0:  # Only consider empty cells
                    # Temporarily place a sensor
                    temp_grid = grid.copy()
                    place_sensor(temp_grid, i, j, coverage_range)
                    covered_count = np.sum(temp_grid)

                    if covered_count > max_covered:
                        max_covered = covered_count
                        best_position = (i, j)

        # Place the best sensor found
        if best_position:
            # the original grid not the tempgrid
            place_sensor(grid, best_position[0], best_position[1], coverage_range)
            sensors_placed.append(best_position)

    return grid, sensors_placed

def visualize_grid(grid, sensors):
    plt.imshow(grid, cmap='Greens')
    plt.colorbar(label='Coverage')
    
    # Mark sensors on the grid
    for sensor in sensors:
        plt.text(sensor[1], sensor[0], 'S', fontsize=12, ha='center', color='red')
    
    plt.title("Sensor Placement on Grid")
    plt.show()

# Parameters
grid_size = 12
sensor_count = 3
coverage_range = 3 # 1 means it can cover 1 cell around

# Run the algorithm
grid, sensors = optimal_sensor_placement(grid_size, sensor_count, coverage_range)
print(grid)
print(sensors)

# Visualize the result
visualize_grid(grid, sensors)
````

## Boundaries Explained

### `max(0, x - coverage_range)`:
- **Purpose**: This ensures that the starting row index does not go below `0`.
- **Explanation**: If the sensor is placed at row `x`, and the `coverage_range` is greater than `x`, the calculated starting index `(x - coverage_range)` would be negative. Since negative indices are invalid for the grid, `max(0, x - coverage_range)` ensures that the starting point is at least `0`.

### `min(size, x + coverage_range + 1)`:
- **Purpose**: This ensures that the ending row index does not exceed the grid size.
- **Explanation**: This calculates the ending row index as `x + coverage_range`. However, if `x + coverage_range` exceeds the maximum index of the grid, it would lead to an out-of-bounds error. The `min(size, x + coverage_range + 1)` ensures that the ending index is capped at the maximum valid index of the grid, which is `size - 1`. The `+ 1` is included because the end of the range in Python slicing is exclusive.

### `max(0, y - coverage_range)`:
- **Purpose**: Similar to the first boundary, this ensures that the starting column index does not go below `0`.
- **Explanation**: This guarantees that the coverage area does not attempt to access a negative column index.

### `min(size, y + coverage_range + 1)`:
- **Purpose**: This ensures that the ending column index does not exceed the grid size.
- **Explanation**: Just like with the row index, this caps the ending column index to avoid going out of bounds when adding coverage.

**Problem:** In a graph where nodes are destinations, find the shortest path that passes through multiple specified points.

````python
import heapq
from itertools import permutations

class Graph:
    def __init__(self):
        self.nodes = {} # The constructor method initializes an empty dictionary called nodes.
        """ 
The Graph class uses a dictionary to represent the adjacency list 
of an undirected graph, where each key is a node and its value is 
another dictionary representing the connected nodes and their edge weights.
        """    
    def add_edge(self, u, v, weight):
        if u not in self.nodes:
            self.nodes[u] = {}
        if v not in self.nodes:
            self.nodes[v] = {}
        self.nodes[u][v] = weight # undirected graph
        self.nodes[v][u] = weight # undirected graph


#  distance_matrix = {node: graph.dijkstra(node) for node in all_nodes}        
    def dijkstra(self, start):
        distances = {node: float('inf') for node in self.nodes}
        # {'A': inf, 'B': inf, 'C': inf, 'D': inf, 'E': inf}

        """ 
creates a dictionary called distances where each node in the graph 
is initialized with a distance of infinity (float('inf')). This represents 
the initial assumption that all nodes are unreachable from the start node 
(hence, infinite distance), except the start node itself, which is later 
set to a distance of 0 with distances[start] = 0        
        """
        distances[start] = 0
        # {'A': 0, 'B': inf, 'C': inf, 'D': inf, 'E': inf}
        priority_queue = [(0, start)] # (distance, node)

        while priority_queue:
            current_distance, current_node = heapq.heappop(priority_queue)
            if current_distance > distances[current_node]:
                continue
            """ 
is a key optimization in Dijkstra's algorithm, known as distance relaxation. 
Here’s how it works and why it’s important:
Purpose of the Check:
This line checks if the current_distance (the distance of the node as 
extracted from the priority queue) is greater than the already recorded 
distances[current_node].
If it is, this means that there’s already a shorter or equal path to this 
node in distances, so further processing of this node with current_distance 
is unnecessary.
            """
            # self.nodes is a dictionary where each key is a node, and its value 
            # is another dictionary that contains neighboring nodes as keys and 
            # their corresponding edge weights as values.
            for neighbor, weight in self.nodes[current_node].items():
                distance = current_distance +weight
                # distance relaxation occurs in Dijkstra's algorithm
                if distance < distances[neighbor]:
                    # If this condition is met, it means the algorithm has 
                    # found a shorter path to the neighbor node.
                    distances[neighbor] = distance
                    heapq.heappush(priority_queue, (distance, neighbor))
        
        return distances

def find_shortest_path(graph, start, destinations):
    min_cost = float('inf')
    best_path = []
    # Include the start node in the destinations list
    all_nodes = [start] + destinations

    # Calculate the distance between each pair of destinations
    distance_matrix = {node: graph.dijkstra(node) for node in all_nodes}
    
    # Check all permutations of the destinations to find the shortest route
    for perm in permutations(destinations):
        """ 
The function generates all possible orders (permutations) of the destination nodes.
For example, if your destinations are ['B', 'C', 'D'], the permutations would be:
['B', 'C', 'D']
['B', 'D', 'C']
['C', 'B', 'D']
['C', 'D', 'B']
['D', 'B', 'C']
['D', 'C', 'B']
        """
        current_cost = 0
        current_path = [start]

        # Add cost from start to the first destination
        current_cost += distance_matrix[start][perm[0]]
        """  
distance_matrix = {
    'A': {'A': 0, 'B': 5, 'C': 8, 'D': 7, 'E': 14},
    'B': {'A': 5, 'B': 0, 'C': 3, 'D': 2, 'E': 9},
    'C': {'A': 8, 'B': 3, 'C': 0, 'D': 1, 'E': 8},
    'D': {'A': 7, 'B': 2, 'C': 1, 'D': 0, 'E': 7},
    'E': {'A': 14, 'B': 9, 'C': 8, 'D': 7, 'E': 0}
}        
        """        
        current_path.append(perm[0])

        # Add costs between subsequent destinations
        for i in range(len(perm) - 1):
            current_cost += distance_matrix[perm[i]][perm[i + 1]]
            current_path.append(perm[i + 1])
        
        # Check if the route is the best found so far
        if current_cost < min_cost:
            min_cost = current_cost
            best_path = current_path

    return min_cost, best_path       

        
# Example usage
graph = Graph()
# Adding edges: (u, v, weight)
edges = [
    ('A', 'B', 5),
    ('A', 'C', 10),
    ('B', 'C', 3),
    ('B', 'D', 2),
    ('C', 'D', 1),
    ('D', 'E', 7),
    ('C', 'E', 8)
]

for u, v, weight in edges:
    graph.add_edge(u, v, weight)

start_node = 'A'
destinations = ['B', 'C', 'D', 'E']

min_cost, best_path = find_shortest_path(graph, start_node, destinations)
print("Minimum Cost:", min_cost)
print("Best Path:", best_path)        
        
````

## Step-by-Step Explanation of Code Behavior

### Graph Initialization
- The `Graph` class initializes an empty dictionary to store nodes and their neighbors with edge weights.
- `add_edge()` adds edges between nodes in both directions (since it’s an undirected graph) with the specified weight.

### Dijkstra’s Algorithm in `dijkstra()`
Dijkstra’s algorithm calculates the shortest distance from the start node to all other nodes in the graph.

It uses:
- A `distances` dictionary to store the shortest known distance from start to each node, initialized to `inf` (except the start node, which is 0).
- A `priority_queue` that stores `(distance, node)` tuples, using `heapq` to keep it ordered by the shortest known distance.

Execution Flow:
1. While there are nodes in the `priority_queue`, the algorithm:
   - Pops the node with the current shortest known distance.
   - Checks each of its neighbors to see if a shorter path exists to that neighbor via the current node.
   - If so, updates the `distances` dictionary and pushes the new distance and neighbor to the priority queue.

Return Value:
- `dijkstra()` returns a dictionary of the shortest distances from the start node to all other nodes in the graph.

### Finding the Shortest Path with `find_shortest_path()`
This function uses a combination of Dijkstra’s algorithm and permutations to find the shortest path that starts at a given node and passes through a specified list of destination nodes.

Steps:
1. **Calculate Pairwise Distances**:
   - First, the function calls `dijkstra()` for each destination node to get the shortest paths between all pairs of destination nodes.
   - This creates a `distance_matrix`, a dictionary storing distances between every node and every other node in `destinations`.

2. **Permutations of Destinations**:
   - It uses `itertools.permutations` to generate all possible orders (routes) in which to visit the destinations.
   - For each permutation, the function calculates the total path cost from the start node, through the permutation of destinations, and finds the minimum cost.

3. **Path Cost Calculation**:
   - The path cost includes:
     - The cost from the start node to the first destination.
     - The costs between subsequent destinations in the permutation.
   - If the path cost for the current permutation is the minimum found so far, it updates `min_cost` and `best_path`.

Return Value:
- The function returns the minimum cost and the best path order to achieve it.

**Battery Management Optimization**

**Problem:** Given a series of battery levels and power usages, maximize travel distance with limited charging stations.

**Application:** Energy management in electric autonomous vehicles.

````python
class ChargingStation:
    def __init__(self, location, charging_rate):
        self.location = location # distance from starting point
        self.charging_rate = charging_rate # charge per hour (KW)
    def __repr__(self):
        return f"Station(Location: {self.location}, Rate: {self.charging_rate})"


class Vehicle:
    def __init__(self, battery_capacity, power_usage):
        self.battery_capacity = battery_capacity # in kw
        self.power_usage = power_usage # in miles per kwh
        self.battery_level = battery_capacity # initially full
    
    def distance_to_empty(self):
        return self.battery_level * self.power_usage # Max distance vehicle can travel


def maximize_travel_distance(vehicle, charging_stations, trip_distance):
    current_position = 0
    travel_distance = 0
    stations_visited = []

    for station in sorted(charging_stations, key=lambda s: s.location):
        distance_to_station = station.location - current_position

        if distance_to_station > vehicle.distance_to_empty():
            break # can not reach the next station

        # update travel distance and position
        travel_distance += distance_to_station
        current_position = station.location

        # Recharge the vehicle at the station
        charging_time = (vehicle.battery_capacity - vehicle.battery_level) / station.charging_rate 
        vehicle.battery_level = vehicle.battery_capacity # fully charged
        stations_visited.append((station, charging_time))

    # check if we can still travel after last charging station
    if current_position < trip_distance:
        remaining_distance = trip_distance - current_position
        if remaining_distance <= vehicle.distance_to_empty():
            travel_distance += remaining_distance  # Reach the destination
        else:
            travel_distance += vehicle.distance_to_empty()  # Max out on battery
    return travel_distance, stations_visited


# Example usage
vehicle = Vehicle(battery_capacity=100, power_usage=4)  # 100 kWh battery, 4 miles/kWh
charging_stations = [
    ChargingStation(location=50, charging_rate=20),  # 20 kW charging
    ChargingStation(location=100, charging_rate=10), # 10 kW charging
    ChargingStation(location=150, charging_rate=15), # 15 kW charging
]
trip_distance = 600  # Total distance to travel
max_distance, stations_visited = maximize_travel_distance(vehicle, charging_stations, trip_distance)

print("Maximum Travel Distance:", max_distance)
print("Charging Stations Visited (Location, Charging Time):", [(station.location, time) for station, time in stations_visited])

````

````python
import heapq
import matplotlib.pyplot as plt
from itertools import permutations

class ChargingStation:
    def __init__(self, location, charging_rate):
        self.location = location  # Node index
        self.charging_rate = charging_rate  # Charge per hour (kW)


class Vehicle:
    def __init__(self, battery_capacity, power_usage):
        self.battery_capacity = battery_capacity  # in kWh
        self.power_usage = power_usage  # in miles per kWh
        self.battery_level = battery_capacity  # initially full

    def distance_to_empty(self):
        return self.battery_level * self.power_usage  # Max distance vehicle can travel


class Graph:
    def __init__(self):
        self.nodes = {}  # The constructor method initializes an empty dictionary called nodes.

    def add_edge(self, u, v, weight):
        if u not in self.nodes:
            self.nodes[u] = {}
        if v not in self.nodes:
            self.nodes[v] = {}
        self.nodes[u][v] = weight  # undirected graph
        self.nodes[v][u] = weight  # undirected graph
    
    def dijkstra(self, start, vehicle):
        distances = {node: float('inf') for node in self.nodes}
        previous_nodes = {node:None for node in self.nodes} # for Tracking the path 
        distances[start] = 0
        priority_queue = [[0, start, vehicle.battery_level]]


        while priority_queue:
            current_distance, current_node, battery_level = heapq.heappop(priority_queue)
            
            if current_distance > distances[current_node]:
                continue

            for neighbour, weight in self.nodes[current_node].items():
                if weight <= vehicle.distance_to_empty(): #check if we can reach 
                    new_battery_level = battery_level - (weight / vehicle.power_usage)
                    new_distance = current_distance + weight # for traking the rest of destinace


                    if new_distance < distances[neighbour]:
                        distances[neighbour] = new_distance
                        previous_nodes[neighbour] = current_node
                        heapq.heappush(priority_queue, (distances[neighbour], neighbour, new_battery_level))
        return distances, previous_nodes

    def reconstruct_path(self, previous_nodes, end_node):
        path = []
        while end_node is not None:
            path.insert(0, end_node)
            end_node = previous_nodes[end_node]
        return path

    def optimize_route(vehicle, graph, charging_stations, trip_distance, start_node=0):
        # start_node = start_node  # starting node
        # get optimal cost and path
        distances, previous_nodes = graph.dijkstra(start_node, vehicle)
        end_node = trip_distance # Distination node
        # if we can reach the destination without charging
        if distances[end_node] < float('inf'): # case distances = 0 this means distances [and node] = inf
            return None, [], [] # Destination unreachable
        
        path_taken = graph.reconstruct_path(previous_nodes, end_node)
        # check if charging stations are needed
        current_battery = vehicle.battery_capacity
        charging_stops = []


        for i in range(1, len(path_taken)):
            segment_distance = graph.nodes[path_taken[i-1]][path_taken[i]]
            battery_needed = segment_distance / vehicle.power_usage
            if current_battery < battery_needed:
                station = path_taken[i-1]
                if station in charging_stations:
                    recharge_needed = vehicle.battery_capacity - current_battery
                    charging_time = recharge_needed / charging_stations[station].charging_rate
                    charging_stops.append((station, charging_time))
                    current_battery = vehicle.battery_capacity  # Recharge battery  
                else:
                    return None, path_taken, charging_stops  # Can't proceed without charging
            current_battery -= battery_needed  # Deduct battery for the segment traveled
        return distances[end_node], path_taken, charging_stops

# Example graph representation
graph = Graph()
# Adding edges: (u, v, weight)
edges = [
    (0, 1, 800),
    (0, 2, 800),
    (1, 3, 400),
    (2, 4, 400),
    (3, 4, 800),
]

for u, v, weight in edges:
    graph.add_edge(u, v, weight)

# Define charging stations and their charging rates
charging_stations = {
    1: ChargingStation(location=1, charging_rate=20),  # Station at node 1
    4: ChargingStation(location=4, charging_rate=10),  # Station at node 4
}

# Create vehicle with battery parameters
vehicle = Vehicle(battery_capacity=200, power_usage=4)  # 100 kWh battery, 4 miles/kWh

# Define the trip distance in terms of the end node
trip_distance = 3  # Destination node



# Optimize route
min_cost, path_taken, charging_stops = optimize_route(vehicle, graph, charging_stations, trip_distance)

# Output results
print("Minimum Fuel Cost to Reach Destination:", min_cost)
print("Route Taken (Nodes):", path_taken)
print("Charging Stations Visited (Location, Charging Time):", charging_stops)
        
````
